**ิฑีผีกีพีฅีฌีธึีฉีตีธึีถีจ** - ิฑีฝีซีถีญึีธีถ ีดีธีฟีฅึีดีกีถ ีคีฅีบึีธึีด ีคีธึึ ีชีกีดีกีถีกีฏ ีนีฅึ ีฏีธึึีถีธึีด ีฝีบีกีฝีฅีฌีซีฝ:

````python
import time
import asyncio
from datetime import datetime

def current_time():
    return datetime.now().strftime("%H:%M:%S")

# ีีซีถีญึีธีถ (blocking) ีดีธีฟีฅึีธึีด
def order_kebab_sync(name, prep_time):
    print(f"{current_time()} - {name}: ีีกีฟีพีซึีธึีด ีฅีด ึีตีกีขีกีข")
    print(f"{current_time()} - ีีกีณีกีผีธีฒ: ิผีกีพ, {prep_time} ีพีกึีฏีตีกีถีซึ ีบีกีฟึีกีฝีฟ ีฏีฌีซีถีซ")
    
    # ิฒีฌีธีฏีกีพีธึีพีกีฎ ีฝีบีกีฝีธึีด
    time.sleep(prep_time)
    
    print(f"{current_time()} - ีีกีณีกีผีธีฒ: {name}, ีฑีฅึ ึีตีกีขีกีข ีบีกีฟึีกีฝีฟ ีง!\n")

    
def main_sync():
    print("๐ฅ ีิปีิฝีีี ีีีิตีีีี")
    print("=" * 50)
    
    start_time = time.time()
    
    # ิตึีฅึ ีกีถีฑ ีบีกีฟีพีซึีธึีด ีฅีถ
    order_kebab_sync("ิทีผีถีฅีฝีฟีธ ีีกึีบีฅึีซ", 5)
    order_kebab_sync("ิฒีฅึีซีดีธีผ", 4)
    order_kebab_sync("ิฑีตีฎีฅีดีถีซีฏ", 7)

    end_time = time.time()
    total_time = end_time - start_time
    
    print(f"\nโฑ๏ธ  ิธีถีคีฐีกีถีธึึ ีชีกีดีกีถีกีฏ: {total_time:.1f} ีพีกีตึีฏีตีกีถ")
    print("ิฒีธีฌีธึีจ ีบีกีฟีพีซึีฅีฌ ีฅีถ ีดีฅีฏ ีกีผ ีดีฅีฏ")

main_sync()
````

````python
# ิฑีฝีซีถีญึีธีถ (non-blocking) ีดีธีฟีฅึีธึีด
async def order_shawarma_async(name, prep_time):
    print(f"{current_time()} - {name}: ีีกีฟีพีซึีธึีด ีฅีด ึีตีกีขีกีข")
    print(f"{current_time()} - ีีกีณีกีผีธีฒ: ิผีกีพ {name}, {prep_time} ีพีกีตึีฏีตีกีถ ีฏีฟึีซ")
    
    # ิฑีฝีซีถีญึีธีถ ีฝีบีกีฝีธึีด (ีกีตีฌ ีฃีธึีฎีธีฒีธึีฉีตีธึีถีถีฅึ ีฏีกึีธีฒ ีฅีถ ีทีกึีธึีถีกีฏีพีฅีฌ)
    await asyncio.sleep(prep_time)
    
    print(f"{current_time()} - ีีกีณีกีผีธีฒ: {name}, ีฑีฅึ ึีตีกีขีกีขีจ ีบีกีฟึีกีฝีฟ ีง!\n")

async def main_async():
    print("๐ ิฑีิปีิฝีีี ีีีิตีีีี")
    print("=" * 50)
    
    start_time = time.time()
    
    # ิฒีธีฌีธึ ีบีกีฟีพีฅึีถีฅึีจ ีฝีฏีฝีพีธึีด ีฅีถ ีฐีกีดีกีชีกีดีกีถีกีฏีตีก
    tasks = [
        order_shawarma_async("ิทีผีถีฅีฝีฟีธ ีีกึีบีฅึีซ", 5),
        order_shawarma_async("ิฒีฅึีซีดีธีผ", 4), 
        order_shawarma_async("ิฑีตีฎีฅีดีถีซีฏ", 7)
    ]
    
    # ีีบีกีฝีธึีด ีฅีถึ ีขีธีฌีธึ ีบีกีฟีพีฅึีถีฅึีซ ีกีพีกึีฟีซีถ
    results = await asyncio.gather(*tasks)
    
    end_time = time.time()
    total_time = end_time - start_time
    
    print(f"\nโฑ๏ธ  ิธีถีคีฐีกีถีธึึ ีชีกีดีกีถีกีฏ: {total_time:.1f} ีพีกีตึีฏีตีกีถ")
    print("ิฒีธีฌีธึ ีบีกีฟีพีฅึีถีฅึีจ ีดีทีกีฏีพีฅีฌ ีฅีถ ีดีซีกีชีกีดีกีถีกีฏ!")
    
    return results

# ินีฅีฝีฟ ีกึีฟีกีฐีกีตีฟีฅีถึ
await main_async()
````

````python
# ๐ก๏ธ Error Handling - ิณีธึีฎีถีกีฏีกีถ ึึีซีถีกีฏีถีฅึ

from fastapi import HTTPException
import random

# ีีกึีฆ ีฝีญีกีฌีถีฅึีซ ึีธึึีกีคึีธึีฉีตีธึีถ
def demonstrate_status_codes():
    """HTTP status codes-ีซ ึีธึึีกีคึีธึีฉีตีธึีถ"""
    
    print("๐ฆ HTTP STATUS CODES - ิตีิตีิฑีีิฑี ีิฑีีีีิฑ")
    print("=" * 55)
    
    scenarios = [
        {
            "situation": "ีีกีณีกีญีธึีคีจ ีฐีกึึีถีธึีด ีง ึีกีถีฏีจ",
            "code": 200,
            "message": "OK - ีีกีถีฏีจ ีขีกึีฅีฐีกีปีธีฒ ีธึีฒีกึีฏีพีฅึ"
        },
        {
            "situation": "ีีธึ ีบีกีฟีพีฅึ ีง ีฝีฟีฅีฒีฎีพีฅีฌ",
            "code": 201,
            "message": "Created - ีีกีฟีพีฅึีจ ีฐีกีปีธีฒีธึีฉีตีกีดีข ีฝีฟีฅีฒีฎีพีฅึ"
        },
        {
            "situation": "ีีกึึีถีธึีด ีฅีถ ีนีฃีธีตีธึีฉีตีธึีถ ีธึีถีฅึีธีฒ ีธึีฟีฅีฌีซึ",
            "code": 404,
            "message": "Not Found - ิฑีตีคีบีซีฝีซ ีธึีฟีฅีฌีซึ ีนีธึีถีฅีถึ"
        },
        {
            "situation": "ีีกีฟีพีซึีธึีด ีฅีถ ึีกีฏ ีชีกีดีฅึีซีถ",
            "code": 403,
            "message": "Forbidden - ีีฅีฝีฟีธึีกีถีจ ึีกีฏ ีง"
        },
        {
            "situation": "ีีญีกีฌ ีฟีพีตีกีฌีถีฅึ ีฅีถ ีธึีฒีกึีฏีฅีฌ",
            "code": 400,
            "message": "Bad Request - ีีฐีกีฝีฏีกึีก ีฑีฅึ ีบีกีฟีพีฅึีจ"
        },
        {
            "situation": "ิฝีธีฐีกีถีธึีธึีด ีฐึีคีฅีฐ ีง",
            "code": 500,
            "message": "Internal Server Error - ีีฅึึีซีถ ีญีถีคีซึ"
        }
    ]
    
    for i, scenario in enumerate(scenarios, 1):
        # Determine status emoji based on code range
        if 200 <= scenario["code"] < 300:
            status_emoji = "โ"
        elif 400 <= scenario["code"] < 600:
            status_emoji = "โ"
        else:
            status_emoji = "๐"
        print(f"\n{i}. {scenario['situation']}")
        print(f"   {status_emoji} {scenario['code']} - {scenario['message']}")

# ีีญีกีฌีถีฅึีซ simulation
class ShawarmaError(Exception):
    """ีีกีธึึีดีก-ีซ ีฝีญีกีฌีถีฅึีซ class"""
    pass

def simulate_order_errors(order_item: str):
    """ีีกีฟีพีฅึีซ ีฝีญีกีฌีถีฅึีซ simulation"""
    
    # ีีกึีขีฅึ ีฝีญีกีฌีถีฅึีซ ึีธึึีกีคึีธึีฉีตีธึีถ
    error_scenarios = {
        "ีฟีกีพีกึีธีพ": {"available": True, "prep_time": 5},
        "ีฐีกีพีธีพ": {"available": True, "prep_time": 3},
        "ีขีกีถีปีกึีฅีฒีฅีถีธีพ": {"available": False, "error": "ิฒีกีถีปีกึีฅีฒีฅีถีจ ีพีฅึีปีกึีฅีฌ ีง"},
        "ีฑีฏีถีธีพ": {"available": False, "error": "ีีธึีฏ ีนีธึีถีฅีถึ"},
        "ีฐีกีฟีธึีฏ": {"available": True, "prep_time": 8, "special_error": "ีีกีฟีธึีฏ ีขีกีฒีกีคึีซีนีจ ีนีฏีก"}
    }
    
    if order_item not in error_scenarios:
        raise HTTPException(status_code=404, detail=f"'{order_item}' ีทีกีธึึีดีก ีนีธึีถีฅีถึ")
    
    scenario = error_scenarios[order_item]
    
    if not scenario["available"]:
        raise HTTPException(status_code=422, detail=scenario["error"])
    
    if "special_error" in scenario and random.choice([True, False]):
        raise HTTPException(status_code=500, detail=scenario["special_error"])
    
    return {"message": f"{order_item} ีทีกีธึึีดีก ีบีกีฟึีกีฝีฟ ีฏีฌีซีถีซ {scenario['prep_time']} ึีธีบีฅีธึีด", 
            "prep_time": scenario["prep_time"]}

# ินีฅีฝีฟ
print("๐ ีิฝิฑิผีิตีิป ีีีีิฑิดีีีินีีีี:")
demonstrate_status_codes()

print("\n" + "="*60)
print("๐งช ีิฑีีิตีีิตีิป ินิตีี:")

test_orders = ["ีฐีกีพีธีพ", "ีขีกีถีปีกึีฅีฒีฅีถีธีพ", "ีฑีฏีถีธีพ", "ีฏีกีฟีพีธีพ"]

for order in test_orders:
    try:
        result = simulate_order_errors(order)
        print(f"โ {order}: {result['message']}")
    except HTTPException as e:
        print(f"โ {order}: {e.status_code} - {e.detail}")
    except Exception as e:
        print(f"๐ฅ {order}: ิฑีถีฝีบีกีฝีฅีฌีซ ีฝีญีกีฌ - {e}")
````
