```python
from pathlib import Path
import pandas as pd

XLSX_PATH = Path("assets", "kargin_internal.xlsx") # os.path.join("assets", "kargin.xlsx")
XLSX_PATH
```

```python
df1 = pd.read_excel(XLSX_PATH)
df2 = pd.read_excel(XLSX_PATH, sheet_name=1)
```

```python
cols = ['titles', 'links', '’á’°’ø ’ø’°÷Ä’°’Æ’æ’°’Æ ’°÷Ä’ø’°’∞’°’µ’ø’∏÷Ç’©’µ’∏÷Ç’∂',
       '’è’•÷Ñ’Ω’ø', '’Ä’´’¥’∂’°’Ø’°’∂ ’§’•÷Ä’°’Ω’°’∂’∂’•÷Ä', '’∞’´’¥’∂’°’Ø’°’∂ ’§’•÷Ä’°’Ω’°’∂’∂’•÷Ä’´ ÷Ñ’°’∂’°’Ø',
       '‘¥’•÷Ä’•÷Ä’´ ’°’∂’∏÷Ç’∂’∂’•÷Ä', '’é’°’µ÷Ä', '‘º’∏÷Ç’Ω’°’æ’∏÷Ä’æ’°’Æ’∏÷Ç’©’µ’∏÷Ç’∂', '‘º’•’¶’∏÷Ç’∂’•÷Ä', '‘±÷Ä’æ’°’Æ ’ß']

df1 = df1[cols]
df1
# df2 = df2[cols]
```

## ‘¥’•÷Ä’°’Ω’°’∂’∂’•÷Ä’´ ÷Ñ’°’∂’°’Ø

```python
df["main_actors_count"].dtype
```

```python
def fix_main_actor(x) -> int:
    try:
        return int(x)
    except Exception:
        return 0

```

```python
df["main_actors_count_int"] = df["main_actors_count"].apply(fix_main_actor)#.value_counts() # masking
```

```python
df["main_actors_count_int"].describe()
```

### fuzzywuzzy

```python
!uv pip install fuzzywuzzy
```

```python
from fuzzywuzzy import fuzz
```

```python
def get_similarity(text_kargin, search_text):
    return fuzz.token_sort_ratio(search_text.lower(), text_kargin)
```

```python
df['similaratiy'] = df["text"].apply(lambda x: get_similarity(x, search_text))
```

```python
df.sort_values("similaratiy", ascending=False)
```

## Youtube-’´÷Å ’´’∂÷Ü’∏

```python
from pytubefix import YouTube 
```

```python
def get_video_info(video_id):
    yt = YouTube(video_id)
    return {
        "title": yt.title,
        "views": yt.views,
        "length": yt.length,
        "author": yt.author,
        "published_at": yt.publish_date
    }
```

```python
df_small = df.head()
```

```python
df_small["youtube_info"] = df_small["links"].apply(get_video_info)
```
