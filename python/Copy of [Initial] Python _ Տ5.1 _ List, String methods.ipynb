## սկիզբ/վերջը ստուգել

```python
a = 'panir.com'
b = 'panir hamov'
```

```python
print(a.startswith('panir')) # a[:5] == "panir"

# հեռացնել
print(a.removeprefix('panir'))
# փոխումա՞ թե չէ
print(a)

a = a.removeprefix('panir')
print(a)
```

```python
print(a.endswith('.com'))
print(b.endswith('.com'))


# հեռացնել
print(a.removesuffix('.com'))
```
