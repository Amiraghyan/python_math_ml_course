```python
# bs4
# Beautiful Soup 4

import requests
from bs4 import BeautifulSoup

URL = "https://www.ysu.am/faculty/516/staff"

response = requests.get(URL)
soup = BeautifulSoup(response.content, "html.parser")

print(soup.prettify())
```

```python
grid = soup.find("div", class_="view-content row")
print(grid)
```

```python
cells =  grid.find_all("div", class_="col-6 col-md-3 views-row")

print(type(cells))
print(cells[0])
```

```python
cells[0].find("a")
```

```python
print(cells[0].find("a").attrs)
cells[0].find("a")["href"]
```

```python
name = cells[0].find("div", class_="user-full-name").text.strip()
```

```python
def clean_name(name):
    return name.strip().replace("\n", " ").replace("  ", " ")

clean_name(name)
```

```python
position = cells[0].find("div", class_="field field--name-field-user-intranet-position field--type-entity-reference field--label-hidden field__item").text.strip()
position
```

```python
from tqdm import tqdm

res = []

for cell in tqdm(cells):
    url = cell.find("a")["href"]
    img = cell.find("img")["src"]
    name = clean_name(cell.find("div", class_="user-full-name").text)
    position = cell.find("div", class_="field field--name-field-user-intranet-position field--type-entity-reference field--label-hidden field__item").text.strip()
    
    res.append({
        "name": name,
        "position": position,
        "url": url,
        "img": img
    })
    
res[0]
```

```python
import pandas as pd

df = pd.DataFrame(res)
df
```

```python
df.isna().sum()
```

```python
assert df.isna().sum().sum() == 0, "There are missing values in the DataFrame"
```

## Per person

```python
import os
from pathlib import Path

BASE_URL = "https://www.ysu.am"
SAVE_DIR = Path("scraping", "htmls_ysu")

os.makedirs(SAVE_DIR, exist_ok=True)

i = 2
name = clean_name(df.iloc[i]["name"])
url_suffix = df.iloc[i]["url"]


url = BASE_URL + url_suffix

response = requests.get(url)
response    
```

```python
print(response.text)
```

```python
filename = f"{SAVE_DIR}/{i}_{clean_name(name)}.html"

with open(filename, "w", encoding="utf-8") as file:
    file.write(response.text)
```

```python
def get_html(i, row):
    print(f"Processing {i}: {row['name']}")
    if i < 10:
        i = f"0{i}"

    name = clean_name(row["name"])
    filename = Path(SAVE_DIR) / f"{i}_{name}.html"

    if filename.exists():
        print(f"File {filename} already exists. Skipping download.")
        return 

    url_suffix = row["url"]
    url = BASE_URL + url_suffix
    
        
    response = requests.get(url)
    response.raise_for_status()  # Ensure we got a successful response
    
    text = response.text
    
    with open(filename, "w", encoding="utf-8") as file:
        file.write(text)
        
    return filename
```

```python
%%time
for i, row in tqdm(df.iterrows(), total=len(df)):
    get_html(i, row)
```

## Parallel processing


```python
from joblib import Parallel, delayed
```

```python
%%time

Parallel(n_jobs=4)( # -1 = all
    delayed(get_html)(i, row) for i, row in df.iterrows()
)
```

## Year of birth

```python
df
```

```python
def get_year_of_birth(soup):
    personal_info = soup.find("div", class_ = "personal-information")
    date_of_birth = personal_info.find("div", class_="fw-bold d-flex field field--name-field-user-date-of-birth field--type-datetime field--label-above")
    date_of_birth_text_list = date_of_birth.text.strip().split("\n")
    title = date_of_birth_text_list[0]
    year = date_of_birth_text_list[1]
    
    
    if title == "‘æ’∂’∂’§’µ’°’∂ ’ø’°÷Ä’•’©’´’æ":
        year_of_birth = year
    else:
        print(f"Title mismatch: {title}. Expected '‘æ’∂’∂’§’µ’°’∂ ’ø’°÷Ä’•’©’´’æ'.")
        year_of_birth = "Unknown"
        
    return year_of_birth
        
```

```python
for i, row in tqdm(df.iterrows(), total=len(df)):
    if i < 10:
        i_file = f"0{i}"
    else:
        i_file = i
    file_name = f"{SAVE_DIR}/{i_file}_{clean_name(row['name'])}.html"
    
    try:
        with open(file_name, "r", encoding="utf-8") as f:
            html = f.read()
    except FileNotFoundError:
        print(f"File {file_name} not found. Skipping {row['name']}.")
        continue
    
    soup = BeautifulSoup(html, "html.parser")
    try:
        year_of_birth = get_year_of_birth(soup)
    except Exception as e:
        print(f"Error processing {i} {row['name']}: {e}")
        year_of_birth = None
    
    df.loc[i, "year"] = year_of_birth
    
```
