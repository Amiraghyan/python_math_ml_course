## Async/Await

- **Սինխրոն ծրագրավորում** - ծրագիրը կատարում է մեկ գործողություն, սպասում է դրա ավարտին, հետո անցնում է հաջորդին:
- **Ասինխրոն ծրագրավորում** - ծրագիրը կարող է սկսել գործողություն, չսպասել դրա ավարտին, և շարունակել այլ գործողություններ:

### Օրինակ

Պատկերացրեք, որ գնում եք «Երևանյան Շաուրմա» (կամ «Արտաշի մոտ», կամ «ՕՍՏ», կամ «Տաշիր», կամ «12 Սեղան», կամ «Արտլանչ», կամ «Սբթայթլ», կամ «KFC», կամ «Համեղ պառառ»):

## 🥙 Սինխրոն մոտեցում (Blocking)

```
Դուք → Վաճառող: "Մեկ շաուրմա՝ հավով, առանց սոխ"
Վաճառող: "Լավ, 5 րոպե սպասեք"

[Դուք կանգնած սպասում եք 5 րոպե, ոչինչ չեք կարող անել]

Վաճառող: "Ահա ձեր շաուրման"
Դուք: "Շնորհակալություն"
[Միայն այժմ կարող եք շարունակել ուրիշ գործեր]
```

## 🔄 Ասինխրոն մոտեցում (Non-blocking)

```
Դուք → Վաճառող: "Մեկ շաուրմա՝ հավով, առանց սոխ"
Վաճառող: "Լավ, 5 րոպե կտևի, ես ձեզ կկանչեմ"

[Դուք շարունակում եք ուրիշ գործեր՝ հեռախոս խոսել, ինստագրամ նայել]
[16 րոպե անց (Լիբանանյան շաուրմայի դեպքում)] 

Վաճառող: "Ձեր շաուրման պատրաստ է!"
Դուք: [վերադառնում եք] "Շնորհակալություն"
```

**Առավելությունը** - Ասինխրոն մոտեցման դեպքում դուք ժամանակ չեք կորցնում սպասելիս:

````python
import time
import asyncio
from datetime import datetime

def current_time():
    return datetime.now().strftime("%H:%M:%S")

# Սինխրոն (blocking) մոտեցում
def order_kebab_sync(name, prep_time):
    print(f"{current_time()} - {name}: Պատվիրում եմ քյաբաբ")
    print(f"{current_time()} - Վաճառող: Լավ, {prep_time} վարկյանից պատրաստ կլինի")
    
    # Բլոկավորված սպասում
    time.sleep(prep_time)
    
    print(f"{current_time()} - Վաճառող: {name}, ձեր քյաբաբ պատրաստ է!\n")

    
def main_sync():
    print("🥙 ՍԻՆԽՐՈՆ ՄՈՏԵՑՈՒՄ")
    print("=" * 50)
    
    start_time = time.time()
    
    # Երեք անձ պատվիրում են
    order_kebab_sync("Էռնեստո Փարպեցի", 5)
    order_kebab_sync("Բերիմոռ", 4)
    order_kebab_sync("Այծեմնիկ", 7)

    end_time = time.time()
    total_time = end_time - start_time
    
    print(f"\n⏱️  Ընդհանուր ժամանակ: {total_time:.1f} վայրկյան")
    print("Բոլորը պատվիրել են մեկ առ մեկ")

main_sync()
````

````python
# Ասինխրոն (non-blocking) մոտեցում
async def order_shawarma_async(name, prep_time):
    print(f"{current_time()} - {name}: Պատվիրում եմ շաուրմա")
    print(f"{current_time()} - Վաճառող: Լավ {name}, {prep_time} վայրկյան կտևի")
    
    # Ասինխրոն սպասում (այլ գործողություններ կարող են շարունակվել)
    await asyncio.sleep(prep_time)
    
    print(f"{current_time()} - Վաճառող: {name}, ձեր շաուրման պատրաստ է!\n")

async def main_async():
    print("🚀 ԱՍԻՆԽՐՈՆ ՄՈՏԵՑՈՒՄ")
    print("=" * 50)
    
    start_time = time.time()
    
    # Բոլոր պատվերները սկսվում են համաժամանակյա
    tasks = [
        order_shawarma_async("Էռնեստո Փարպեցի", 5),
        order_shawarma_async("Բերիմոռ", 4), 
        order_shawarma_async("Այծեմնիկ", 7)
    ]
    
    # Սպասում ենք բոլոր պատվերների ավարտին
    results = await asyncio.gather(*tasks)
    
    end_time = time.time()
    total_time = end_time - start_time
    
    print(f"\n⏱️  Ընդհանուր ժամանակ: {total_time:.1f} վայրկյան")
    print("Բոլոր պատվերները մշակվել են միաժամանակ!")
    
    return results

# Թեստ արտահայտենք
await main_async()
````

````python
await order_shawarma_async("Էռնեստո Փարպեցի", 5)
````

# 🌐 HTTP և Request Types

## Ի՞նչ է HTTP-ն

**HTTP (HyperText Transfer Protocol)** - ժամանակակից ինտերնետի հիմքն է: Համարյա ցանկացած կայք, բջջային հավելված կամ API այս պրոտոկոլով է աշխատում:

### 🥙 HTTP-ն «Երևանյան Շաուրմա»-ի օրինակով

Պատկերացրեք HTTP-ն որպես **ռեստորանի պատվերների համակարգ**:

```
Հաճախորդ (Client) ←→ Սերվեր (Server)
     ↓                    ↑
"Request"              "Response"
(Հարցում)              (Պատասխան)
```

### 📋 HTTP Request-ի կառուցվածքը

```http

}
```

# 🔄 HTTP Methods - Երևանյան Շաուրմա օրինակներով

## 📚 Հիմնական հասկացություններ

**Safe (Անվտանգ)** - HTTP մեթոդը համարվում է անվտանգ, եթե այն չի փոխում սերվերի վիճակը։ 
Նման մեթոդները կարող են կանչվել առանց վախենալու, որ կփոխեն տվյալները:

**Idempotent** - HTTP մեթոդը համարվում է idempotent, եթե նույն հարցումը մի քանի անգամ 
ուղարկելուց արդյունքը չի փոխվում։ Այսինքն՝ 1 անգամ կամ 10 անգամ կանչելը նույն արդյունքն է տալիս:

## 🌐 HTTP Methods-ի ցուցադրություն

### GET - Տեղեկություն ստանալ
- **Նկարագրություն**: Տեղեկություն ստանալ (չփոխել ոչինչ)
- **Օրինակ**: Մենյուն նայել
- **🥙 Շաուրմա օրինակ**: «Հավի իքիբիր ունե՞ք», «Ի՞նչքան ա»
- **⚡ Անվտանգ**: Այո
- **🔄 Idempotent**: Այո

### POST - Նոր բան ստեղծել
- **Նկարագրություն**: Նոր բան ստեղծել
- **Օրինակ**: Նոր պատվեր տալ
- **🥙 Շաուրմա օրինակ**: «Ինձ մի հատ հավի շաուրմա խնդրում եմ»
- **⚡ Անվտանգ**: Ոչ
- **🔄 Idempotent**: Ոչ

### PUT - Ամբողջությամբ փոխարինել/թարմացնել
- **Նկարագրություն**: Ամբողջությամբ փոխարինել/թարմացնել
- **Օրինակ**: Պատվերը ամբողջությամբ փոխել
- **🥙 Շաուրմա օրինակ**: «Չեղարկեք հավը, ուզում եմ տավարով»
- **⚡ Անվտանգ**: Ոչ
- **🔄 Idempotent**: Այո

### PATCH - Մասնակիորեն թարմացնել
- **Նկարագրություն**: Մասնակիորեն թարմացնել
- **Օրինակ**: Միայն մեկ բաղադրիչ փոխել
- **🥙 Շաուրմա օրինակ**: «Խնդրում եմ առանց սոխի»
- **⚡ Անվտանգ**: Ոչ
- **🔄 Idempotent**: Ոչ

### DELETE - Ջնջել
- **Նկարագրություն**: Ջնջել
- **Օրինակ**: Պատվերը չեղարկել
- **🥙 Շաուրմա օրինակ**: «Չեղարկում եմ իմ պատվերը»
- **⚡ Անվտանգ**: Ոչ
- **🔄 Idempotent**: Այո

## 📊 Status Codes - Ռեստորանի լեզվով

HTTP Status Code-երը բաժանվում են 5 խմբի:

### 1️⃣ 1xx - Informational (Տեղեկություն)
- **100 Continue** - "Ստացա հարցումը, շարունակեք խոսել, լսում եմ"

### 2️⃣ 2xx - Success (Հաջողություն) ✅
- **200 OK** - "Ամեն ինչ լավ է"
- **201 Created** - "Պատվերը ստեղծվեց"
- **204 No Content** - "Կատարվեց, բայց ուղարկելու բան չկա"

### 3️⃣ 3xx - Redirection (Վերահղում) 🔄
- **301 Moved Permanently** - "Ռեստորանը տեղափոխվել է մեկ այլ հասցե"
- **302 Found** - "Ժամանակավորապես աշխատում ենք այլ հասցեում"

### 4️⃣ 4xx - Client Error (Client-ի սխալ) ❌
- **400 Bad Request** - "Չհասկացա ձեր պատվերը"
- **401 Unauthorized** - "Նույնականացում չեք անցել"
- **403 Forbidden** - "Հասանելի չէ ձեզ համար"
- **404 Not Found** - "Այդպիսի ուտելիք չունենք"
- **418 I'm a teapot** - "Ես թեյնիկ եմ (տես ներքևում)"
- **422 Unprocessable Entity** - "Հասկանալի է, բայց կատարել չեմ կարող"

### 5️⃣ 5xx - Server Error (Սերվերի սխալ) 💥
- **500 Internal Server Error** - "Խոհանոցում խնդիր կա"
- **502 Bad Gateway** - "Մատակարարը չի պատասխանում"
- **503 Service Unavailable** - "Ժամանակավորապես փակ ենք"

*418 I'm a teapot (RFC 2324, RFC 7168)*

The RFC specifies this code should be returned by teapots requested to brew coffee.[19] This HTTP status is used as an Easter egg in some websites, such as Google.com's "I'm a teapot" easter egg.[20][21][22] Sometimes, this status code is also used as a response to a blocked request, instead of the more appropriate 403 Forbidden.[23][24]

````python
# 🛡️ Error Handling - Գործնական օրինակներ

from fastapi import HTTPException
import random

# Պարզ սխալների ցուցադրություն
def demonstrate_status_codes():
    """HTTP status codes-ի ցուցադրություն"""
    
    print("🚦 HTTP STATUS CODES - ԵՐԵՎԱՆՅԱՆ ՇԱՈՒՐՄԱ")
    print("=" * 55)
    
    scenarios = [
        {
            "situation": "Հաճախորդը հարցնում է ցանկը",
            "code": 200,
            "message": "OK - Ցանկը բարեհաջող ուղարկվեց"
        },
        {
            "situation": "Նոր պատվեր է ստեղծվել",
            "code": 201,
            "message": "Created - Պատվերը հաջողությամբ ստեղծվեց"
        },
        {
            "situation": "Հարցնում են չգոյություն ունեցող ուտելիք",
            "code": 404,
            "message": "Not Found - Այդպիսի ուտելիք չունենք"
        },
        {
            "situation": "Պատվիրում են փակ ժամերին",
            "code": 403,
            "message": "Forbidden - Ռեստորանը փակ է"
        },
        {
            "situation": "Սխալ տվյալներ են ուղարկել",
            "code": 400,
            "message": "Bad Request - Չհասկացա ձեր պատվերը"
        },
        {
            "situation": "Խոհանոցում հրդեհ է",
            "code": 500,
            "message": "Internal Server Error - Ներքին խնդիր"
        }
    ]
    
    for i, scenario in enumerate(scenarios, 1):
        # Determine status emoji based on code range
        if 200 <= scenario["code"] < 300:
            status_emoji = "✅"
        elif 400 <= scenario["code"] < 600:
            status_emoji = "❌"
        else:
            status_emoji = "🔄"
        print(f"\n{i}. {scenario['situation']}")
        print(f"   {status_emoji} {scenario['code']} - {scenario['message']}")

# Սխալների simulation
class ShawarmaError(Exception):
    """Շաուրմա-ի սխալների class"""
    pass

def simulate_order_errors(order_item: str):
    """Պատվերի սխալների simulation"""
    
    # Տարբեր սխալների ցուցադրություն
    error_scenarios = {
        "տավարով": {"available": True, "prep_time": 5},
        "հավով": {"available": True, "prep_time": 3},
        "բանջարեղենով": {"available": False, "error": "Բանջարեղենը վերջացել է"},
        "ձկնով": {"available": False, "error": "Ձուկ չունենք"},
        "հատուկ": {"available": True, "prep_time": 8, "special_error": "Հատուկ բաղադրիչը չկա"}
    }
    
    if order_item not in error_scenarios:
        raise HTTPException(status_code=404, detail=f"'{order_item}' շաուրմա չունենք")
    
    scenario = error_scenarios[order_item]
    
    if not scenario["available"]:
        raise HTTPException(status_code=422, detail=scenario["error"])
    
    if "special_error" in scenario and random.choice([True, False]):
        raise HTTPException(status_code=500, detail=scenario["special_error"])
    
    return {"message": f"{order_item} շաուրմա պատրաստ կլինի {scenario['prep_time']} րոպեում", 
            "prep_time": scenario["prep_time"]}

# Թեստ
print("📋 ՍԽԱԼՆԵՐԻ ՑՈՒՑԱԴՐՈՒԹՅՈՒՆ:")
demonstrate_status_codes()

print("\n" + "="*60)
print("🧪 ՊԱՏՎԵՐՆԵՐԻ ԹԵՍՏ:")

test_orders = ["հավով", "բանջարեղենով", "ձկնով", "կատվով"]

for order in test_orders:
    try:
        result = simulate_order_errors(order)
        print(f"✅ {order}: {result['message']}")
    except HTTPException as e:
        print(f"❌ {order}: {e.status_code} - {e.detail}")
    except Exception as e:
        print(f"💥 {order}: Անսպասելի սխալ - {e}")
````

````python
# 🚀 Ամբողջական FastAPI օրինակ - Երևանյան Շաուրմա

# Ցանկային այս օրինակը աշխատացնելու համար պետք է տեղադրել FastAPI:
# pip install fastapi uvicorn

from pydantic import BaseModel, validator
from typing import Optional, List
from datetime import datetime
import asyncio

# --- ՏՎՅԱԼՆԵՐԻ ՄՈԴԵԼՆԵՐ ---

class ShawarmaItem(BaseModel):
    name: str
    price: int
    available: bool = True
    prep_time: int  # րոպեներ

class OrderCreate(BaseModel):
    customer_name: str
    items: List[str]
    special_requests: Optional[str] = ""
    
    @validator('customer_name')
    def validate_name(cls, v):
        if len(v.strip()) < 2:
            raise ValueError('Անունը պետք է լինի նվազագույնը 2 տառ')
        return v.strip()

class Order(BaseModel):
    id: int
    customer_name: str
    items: List[str]
    total_price: int
    status: str
    created_at: datetime
    estimated_time: int

# --- "ՏՎՅԱԼՆԵՐԻ ԲԱԶԱ" ---
menu_items = {
    "հավով": ShawarmaItem(name="հավով", price=1500, prep_time=3),
    "տավարով": ShawarmaItem(name="տավարով", price=1800, prep_time=4),
    "բանջարեղենով": ShawarmaItem(name="բանջարեղենով", price=1200, prep_time=2),
    "հատուկ": ShawarmaItem(name="հատուկ", price=2200, prep_time=6)
}

orders_storage = {}
next_order_id = 1

# --- ՕԺԱՆԴԱԿ ՖՈՒՆԿՑԻԱՆԵՐ ---

async def validate_menu_items(items: List[str]) -> None:
    """Ստուգել՝ արդյոք պատվիրված ապրանքները կան ցանկում"""
    for item in items:
        if item not in menu_items:
            raise HTTPException(
                status_code=404, 
                detail=f"'{item}' շաուրմա մենք չունենք: Առկա տարբերակներ՝ {list(menu_items.keys())}"
            )
        if not menu_items[item].available:
            raise HTTPException(
                status_code=422,
                detail=f"'{item}' շաուրման ժամանակավորապես մատչելի չէ"
            )

def calculate_order_total(items: List[str]) -> tuple[int, int]:
    """Հաշվել պատվերի գումարը և պատրաստման ժամանակը"""
    total_price = sum(menu_items[item].price for item in items)
    total_time = max(menu_items[item].prep_time for item in items)  # Ամենաերկար ժամանակ
    return total_price, total_time

# --- FASTAPI ENDPOINTS SIMULATION ---

async def simulate_get_menu():
    """GET /menu - Ցանկի ստացում"""
    return {
        "status": "success",
        "menu": {name: {"price": item.price, "available": item.available} 
                for name, item in menu_items.items()}
    }

async def simulate_create_order(order_data: OrderCreate):
    """POST /orders - Նոր պատվեր ստեղծում"""
    global next_order_id
    
    # Validation
    await validate_menu_items(order_data.items)
    
    # Calculation
    total_price, prep_time = calculate_order_total(order_data.items)
    
    # Create order
    new_order = Order(
        id=next_order_id,
        customer_name=order_data.customer_name,
        items=order_data.items,
        total_price=total_price,
        status="գործընթաց",
        created_at=datetime.now(),
        estimated_time=prep_time
    )
    
    orders_storage[next_order_id] = new_order
    order_id = next_order_id
    next_order_id += 1
    
    # Simulate cooking
    print(f"🥙 Պատրաստում եմ պատվեր #{order_id} ({order_data.customer_name})")
    await asyncio.sleep(1)  # Simulation
    
    return {
        "status": "created",
        "order": new_order,
        "message": f"Պատվեր #{order_id} ստեղծվեց: Պատրաստ կլինի {prep_time} րոպեում"
    }

async def simulate_get_order(order_id: int):
    """GET /orders/{order_id} - Մեկ պատվերի տվյալներ"""
    if order_id not in orders_storage:
        raise HTTPException(status_code=404, detail=f"Պատվեր #{order_id} չի գտնվել")
    
    return {"status": "success", "order": orders_storage[order_id]}

async def simulate_update_order(order_id: int, new_items: List[str]):
    """PUT /orders/{order_id} - Պատվերը փոխել"""
    if order_id not in orders_storage:
        raise HTTPException(status_code=404, detail=f"Պատվեր #{order_id} չի գտնվել")
    
    order = orders_storage[order_id]
    if order.status != "գործընթաց":
        raise HTTPException(status_code=422, detail="Պատրաստ պատվերը չի կարելի փոխել")
    
    await validate_menu_items(new_items)
    total_price, prep_time = calculate_order_total(new_items)
    
    order.items = new_items
    order.total_price = total_price
    order.estimated_time = prep_time
    
    return {"status": "updated", "order": order}

async def simulate_cancel_order(order_id: int):
    """DELETE /orders/{order_id} - Պատվերը չեղարկել"""
    if order_id not in orders_storage:
        raise HTTPException(status_code=404, detail=f"Պատվեր #{order_id} չի գտնվել")
    
    order = orders_storage[order_id]
    if order.status == "պատրաստ":
        raise HTTPException(status_code=422, detail="Պատրաստ պատվերը չի կարելի չեղարկել")
    
    del orders_storage[order_id]
    return {"status": "cancelled", "message": f"Պատվեր #{order_id} չեղարկվեց"}

# --- ՏԵՍՏԱՎՈՐՈՒՄ ---

async def test_api_endpoints():
    """API endpoints-ի թեստավորում"""
    print("🧪 FASTAPI ENDPOINTS ԹԵՍՏ")
    print("=" * 50)
    
    try:
        # 1. GET Menu
        print("\n1️⃣ GET /menu")
        menu_result = await simulate_get_menu()
        print(f"✅ Ցանկը ստացվեց: {len(menu_result['menu'])} տարբերակ")
        
        # 2. POST Order
        print("\n2️⃣ POST /orders")
        order_data = OrderCreate(
            customer_name="Արամ Պետրոսյան", 
            items=["հավով", "տավարով"],
            special_requests="Առանց սոխի"
        )
        order_result = await simulate_create_order(order_data)
        order_id = order_result["order"].id
        print(f"✅ {order_result['message']}")
        
        # 3. GET Specific Order
        print("\n3️⃣ GET /orders/{order_id}")
        get_result = await simulate_get_order(order_id)
        print(f"✅ Պատվեր #{order_id} գտնվեց")
        
        # 4. PUT Update Order
        print("\n4️⃣ PUT /orders/{order_id}")
        update_result = await simulate_update_order(order_id, ["հատուկ"])
        print(f"✅ Պատվերը թարմացվեց")
        
        # 5. Փորձ - սխալ պատվեր
        print("\n5️⃣ POST /orders (սխալ)")
        try:
            bad_order = OrderCreate(customer_name="Գագիկ", items=["կատվով"])
            await simulate_create_order(bad_order)
        except HTTPException as e:
            print(f"❌ Ակնկալվող սխալ: {e.status_code} - {e.detail}")
        
        # 6. DELETE Cancel Order
        print("\n6️⃣ DELETE /orders/{order_id}")
        cancel_result = await simulate_cancel_order(order_id)
        print(f"✅ {cancel_result['message']}")
        
    except Exception as e:
        print(f"💥 Անսպասելի սխալ: {e}")

# Թեստ
await test_api_endpoints()
````

````python
# 🧪 API-ների թեստավորում

import requests
import json
import sys
import os

def test_api_endpoint(base_url, endpoint, method="GET", data=None):
    """API endpoint-ի թեստավորման ֆունկցիա"""
    url = f"{base_url}{endpoint}"
    
    try:
        if method == "GET":
            response = requests.get(url, timeout=5)
        elif method == "POST":
            headers = {"Content-Type": "application/json"}
            response = requests.post(url, json=data, headers=headers, timeout=5)
        else:
            return f"❌ Անակնկալ method: {method}"
        
        if response.status_code == 200:
            return f"✅ {method} {endpoint}: {response.status_code}"
        elif response.status_code == 201:
            return f"✅ {method} {endpoint}: {response.status_code} - Created"
        else:
            return f"⚠️ {method} {endpoint}: {response.status_code} - {response.text[:100]}"
    
    except requests.exceptions.ConnectionError:
        return f"❌ {method} {endpoint}: Սերվերը չի աշխատում"
    except requests.exceptions.Timeout:
        return f"⏰ {method} {endpoint}: Timeout"
    except Exception as e:
        return f"💥 {method} {endpoint}: {str(e)}"

def test_apis():
    """FastAPI և Flask API-ների թեստավորում"""
    
    print("🧪 API ENDPOINTS ԹԵՍՏ")
    print("=" * 50)
    
    # API bases
    fastapi_base = "http://localhost:8000"
    flask_base = "http://localhost:5000"
    
    # Test data
    test_order = {
        "customer_name": "Արամ Թեստական",
        "items": ["հավով", "բանջարեղենով"],
        "special_requests": "Առանց սոխի"
    }
    
    print("\n🚀 FastAPI թեստ (Port 8000)")
    print("-" * 30)
    
    fastapi_tests = [
        (fastapi_base, "/", "GET", None),
        (fastapi_base, "/menu", "GET", None),
        (fastapi_base, "/orders", "POST", test_order),
        (fastapi_base, "/orders/1", "GET", None),
        (fastapi_base, "/orders", "GET", None),
    ]
    
    for base, endpoint, method, data in fastapi_tests:
        result = test_api_endpoint(base, endpoint, method, data)
        print(f"  {result}")
    
    print("\n🌶️ Flask թեստ (Port 5000)")
    print("-" * 30)
    
    flask_tests = [
        (flask_base, "/", "GET", None),
        (flask_base, "/menu", "GET", None),
        (flask_base, "/orders", "POST", test_order),
        (flask_base, "/orders/1", "GET", None),
        (flask_base, "/orders", "GET", None),
    ]
    
    for base, endpoint, method, data in flask_tests:
        result = test_api_endpoint(base, endpoint, method, data)
        print(f"  {result}")
    
    print(f"\n💡 Նկատառում:")
    print(f"   - Գործարկեք API-ները առանձին terminal-ներում")
    print(f"   - FastAPI: cd apis && python fastapi_shawarma.py")
    print(f"   - Flask:   cd apis && python flask_shawarma.py")

# Թեստ
test_apis()
````

````python
# 🔧 FastAPI տեղադրում և կառուցվածք

# 1. Տեղադրում - Terminal-ում գործարկել:
# pip install fastapi uvicorn python-multipart

# 2. Պարտադիր imports
try:
    from fastapi import FastAPI, HTTPException, Depends, Query, Path
    from fastapi.responses import JSONResponse
    from pydantic import BaseModel, Field
    from typing import Optional, List, Dict, Any
    from datetime import datetime, timedelta
    from enum import Enum
    import asyncio
    import json
    print("✅ FastAPI imports բարեհաջող բեռնավորվեցին")
except ImportError as e:
    print(f"❌ Import Error: {e}")
    print("💡 Խորհուրդ: pip install fastapi uvicorn")

# 3. FastAPI application ստեղծում
app = FastAPI(
    title="🥙 Երևանյան Շաուրմա API",
    description="Ռեստորանի պատվերակման համակարգ",
    version="1.0.0",
    docs_url="/docs",  # Swagger UI
    redoc_url="/redoc"  # ReDoc alternative docs
)

print("🚀 FastAPI application ստեղծվեց")

# 4. Enum classes
class ShawarmaType(str, Enum):
    CHICKEN = "հավով"
    BEEF = "տավարով"
    VEGGIE = "բանջարեղենով"
    SPECIAL = "հատուկ"
    FISH = "ձկնով"

class OrderStatus(str, Enum):
    PENDING = "սպասվող"
    CONFIRMED = "հաստատված"
    PREPARING = "պատրաստվում"
    READY = "պատրաստ"
    DELIVERED = "առաքված"
    CANCELLED = "չեղարկված"

# 5. Pydantic Models (v2 compatible)
class MenuItem(BaseModel):
    name: str = Field(..., description="Ուտելիքի անվանությունը")
    price: int = Field(..., gt=0, description="Գինը դրամով")
    description: str = Field("", description="Նկարագրությունը")
    available: bool = Field(True, description="Հասանելիությունը")
    prep_time: int = Field(..., gt=0, le=30, description="Պատրաստման ժամանակը րոպեներով")
    category: str = Field("shawarma", description="Կատեգորիան")
    
    model_config = {
        "json_schema_extra": {
            "example": {
                "name": "հավով շաուրմա",
                "price": 1500,
                "description": "Հավի միս, բանջարեղեն, սոուս",
                "available": True,
                "prep_time": 5,
                "category": "shawarma"
            }
        }
    }

class OrderItem(BaseModel):
    item_name: str
    quantity: int = Field(..., gt=0, le=10)
    special_requests: Optional[str] = ""

class OrderCreate(BaseModel):
    customer_name: str = Field(..., min_length=2, max_length=50)
    phone: str = Field(..., pattern=r"^\+374\d{8}$")
    items: List[OrderItem] = Field(..., min_length=1)
    delivery_address: Optional[str] = None
    payment_method: str = Field("cash", pattern="^(cash|card|online)$")

class OrderResponse(BaseModel):
    id: int
    customer_name: str
    phone: str
    items: List[OrderItem]
    total_price: int
    status: OrderStatus
    created_at: datetime
    estimated_delivery: Optional[datetime] = None

print("📝 Տվյալների մոդելները պատրաստ են")
````

````python
# 🗄️ "Տվյալների բազա" և հիմնական գործառույթներ

# Պարզ in-memory "database"
class ShawarmaDatabase:
    def __init__(self):
        self.menu: Dict[str, MenuItem] = {}
        self.orders: Dict[int, OrderResponse] = {}
        self.next_order_id = 1
        self._init_menu()
    
    def _init_menu(self):
        """Ցանկի տվյալների ինիցիալիզացիա"""
        menu_items = [
            MenuItem(
                name="հավով շաուրմա",
                price=1500,
                description="Հրապարակված հավի միս, մանր կտրած բանջարեղեն, սոուս",
                prep_time=5,
                category="շաուրմա"
            ),
            MenuItem(
                name="տավարով շաուրմա", 
                price=1800,
                description="Քիչ պատրաստված տավարի միս, բանջարեղեն, հատուկ սոուս",
                prep_time=6,
                category="շաուրմա"
            ),
            MenuItem(
                name="բանջարեղենով շաուրմա",
                price=1200,
                description="Տարբեր բանջարեղեն, հումուս, սոուս",
                prep_time=4,
                category="շաուրմա"
            ),
            MenuItem(
                name="հատուկ շաուրմա",
                price=2200,
                description="Հավ+տավար, հատուկ սոուս, կակաչ",
                prep_time=8,
                category="շաուրմա"
            ),
            MenuItem(
                name="կոլա",
                price=400,
                description="Coca-Cola 0.5լ",
                prep_time=1,
                category="ըմպելիք"
            ),
            MenuItem(
                name="այրան",
                price=300,
                description="Տանական ախկածիր ռիային",
                prep_time=1,
                category="ըմպելիք"
            )
        ]
        
        for item in menu_items:
            self.menu[item.name] = item
    
    def get_menu(self) -> Dict[str, MenuItem]:
        """Ցանկը ստանալ"""
        return self.menu
    
    def get_available_menu(self) -> Dict[str, MenuItem]:
        """Միայն հասանելի ապրանքները"""
        return {name: item for name, item in self.menu.items() if item.available}
    
    def create_order(self, order_data: OrderCreate) -> OrderResponse:
        """Նոր պատվեր ստեղծել"""
        # Ստուգել ապրանքների առկայությունը
        for item in order_data.items:
            if item.item_name not in self.menu:
                raise HTTPException(
                    status_code=404,
                    detail=f"'{item.item_name}' ցանկում չկա"
                )
            if not self.menu[item.item_name].available:
                raise HTTPException(
                    status_code=422,
                    detail=f"'{item.item_name}' ժամանակավորապես մատչելի չէ"
                )
        
        # Գումարի հաշվարկ
        total_price = sum(
            self.menu[item.item_name].price * item.quantity 
            for item in order_data.items
        )
        
        # Ավարտման ժամանակի գնահատում
        max_prep_time = max(
            self.menu[item.item_name].prep_time 
            for item in order_data.items
        )
        estimated_delivery = datetime.now() + timedelta(minutes=max_prep_time + 10)
        
        # Պատվեր ստեղծել
        order = OrderResponse(
            id=self.next_order_id,
            customer_name=order_data.customer_name,
            phone=order_data.phone,
            items=order_data.items,
            total_price=total_price,
            status=OrderStatus.PENDING,
            created_at=datetime.now(),
            estimated_delivery=estimated_delivery
        )
        
        self.orders[self.next_order_id] = order
        self.next_order_id += 1
        
        return order
    
    def get_order(self, order_id: int) -> OrderResponse:
        """Պատվեր գտնել"""
        if order_id not in self.orders:
            raise HTTPException(
                status_code=404,
                detail=f"Պատվեր #{order_id} չի գտնվել"
            )
        return self.orders[order_id]
    
    def update_order_status(self, order_id: int, new_status: OrderStatus) -> OrderResponse:
        """Պատվերի կարգավիճակ փոխել"""
        order = self.get_order(order_id)
        order.status = new_status
        return order
    
    def cancel_order(self, order_id: int) -> bool:
        """Պատվեր չեղարկել"""
        order = self.get_order(order_id)
        if order.status in [OrderStatus.DELIVERED, OrderStatus.CANCELLED]:
            raise HTTPException(
                status_code=422,
                detail=f"Պատվեր #{order_id} հնարավոր չէ չեղարկել: Կարգավիճակ՝ {order.status.value}"
            )
        
        order.status = OrderStatus.CANCELLED
        return True

# Գլոբալ database instance
db = ShawarmaDatabase()

print("🗄️ Տվյալների բազան պատրաստ է")
print(f"📋 Ցանկում {len(db.menu)} ապրանք կա")
print(f"🛒 Պատվերներ: {len(db.orders)}")
````

````python
# 🛣️ FastAPI Routes (Endpoints)

# ROOT endpoint
@app.get("/", tags=["Գլխավոր"])
async def root():
    """API-ի գլխավոր էջ"""
    return {
        "message": "🥙 Բարի գալուստ Երևանյան Շաուրմա API",
        "version": "1.0.0",
        "docs": "/docs",
        "status": "կենդանի"
    }

# MENU endpoints
@app.get("/menu", response_model=Dict[str, MenuItem], tags=["Ցանկ"])
async def get_menu(
    available_only: bool = Query(False, description="Միայն հասանելի ապրանքները")
):
    """Ցանկի ստացում"""
    if available_only:
        return db.get_available_menu()
    return db.get_menu()

@app.get("/menu/{item_name}", response_model=MenuItem, tags=["Ցանկ"])
async def get_menu_item(item_name: str = Path(..., description="Ապրանքի անվանությունը")):
    """Մեկ ապրանքի մանրամասներ"""
    if item_name not in db.menu:
        raise HTTPException(
            status_code=404,
            detail=f"'{item_name}' ցանկում չի գտնվել"
        )
    return db.menu[item_name]

# ORDER endpoints
@app.post("/orders", response_model=OrderResponse, status_code=201, tags=["Պատվերներ"])
async def create_order(order: OrderCreate):
    """Նոր պատվեր ստեղծել"""
    try:
        new_order = db.create_order(order)
        
        # Ասինխրոն պատրաստման սիմուլյացիա
        asyncio.create_task(simulate_order_preparation(new_order.id))
        
        return new_order
    except Exception as e:
        if isinstance(e, HTTPException):
            raise e
        raise HTTPException(status_code=500, detail=f"Ներքին սխալ: {str(e)}")

@app.get("/orders/{order_id}", response_model=OrderResponse, tags=["Պատվերներ"])
async def get_order(order_id: int = Path(..., description="Պատվերի համարը")):
    """Պատվերի կարգավիճակ ստուգել"""
    return db.get_order(order_id)

@app.get("/orders", response_model=List[OrderResponse], tags=["Պատվերներ"])
async def get_all_orders(
    status: Optional[OrderStatus] = Query(None, description="Կարգավիճակով ֆիլտր"),
    limit: int = Query(10, ge=1, le=100, description="Առավելագույն քանակ")
):
    """Բոլոր պատվերները"""
    orders = list(db.orders.values())
    
    if status:
        orders = [order for order in orders if order.status == status]
    
    # Վերջին պատվերները վերևում
    orders.sort(key=lambda x: x.created_at, reverse=True)
    
    return orders[:limit]

@app.put("/orders/{order_id}/status", response_model=OrderResponse, tags=["Պատվերներ"])
async def update_order_status(
    order_id: int = Path(..., description="Պատվերի համարը"),
    new_status: OrderStatus = Query(..., description="Նոր կարգավիճակ")
):
    """Պատվերի կարգավիճակ փոխել"""
    return db.update_order_status(order_id, new_status)

@app.delete("/orders/{order_id}", tags=["Պատվերներ"])
async def cancel_order(order_id: int = Path(..., description="Պատվերի համարը")):
    """Պատվեր չեղարկել"""
    success = db.cancel_order(order_id)
    if success:
        return {"message": f"Պատվեր #{order_id} չեղարկվեց", "status": "success"}
    
# STATISTICS endpoint
@app.get("/stats", tags=["Վիճակագրություն"])
async def get_statistics():
    """Հիմնական վիճակագրություն"""
    orders = list(db.orders.values())
    
    total_orders = len(orders)
    total_revenue = sum(order.total_price for order in orders if order.status != OrderStatus.CANCELLED)
    
    status_counts = {}
    for status in OrderStatus:
        status_counts[status.value] = len([o for o in orders if o.status == status])
    
    popular_items = {}
    for order in orders:
        for item in order.items:
            if item.item_name not in popular_items:
                popular_items[item.item_name] = 0
            popular_items[item.item_name] += item.quantity
    
    return {
        "total_orders": total_orders,
        "total_revenue": total_revenue,
        "status_breakdown": status_counts,
        "popular_items": dict(sorted(popular_items.items(), key=lambda x: x[1], reverse=True)[:5])
    }

# Օժանդակ ֆունկցիա - պատվերի պատրաստման սիմուլյացիա
async def simulate_order_preparation(order_id: int):
    """Պատվերի պատրաստման ասինխրոն սիմուլյացիա"""
    try:
        # Հաստատել
        await asyncio.sleep(2)
        db.update_order_status(order_id, OrderStatus.CONFIRMED)
        
        # Պատրաստել
        await asyncio.sleep(3)
        db.update_order_status(order_id, OrderStatus.PREPARING)
        
        # Պատրաստ
        order = db.get_order(order_id)
        prep_times = [db.menu[item.item_name].prep_time for item in order.items]
        total_prep_time = max(prep_times) if prep_times else 5
        
        await asyncio.sleep(total_prep_time)
        db.update_order_status(order_id, OrderStatus.READY)
        
        # Առաքել (օրինակ)
        await asyncio.sleep(10)
        db.update_order_status(order_id, OrderStatus.DELIVERED)
        
    except Exception as e:
        print(f"Սխալ պատվեր #{order_id} պատրաստելիս: {e}")

print("🛣️ FastAPI routes պատրաստ են")
print("📊 Endpoints:")
print("   GET  /         - Գլխավոր էջ")
print("   GET  /menu     - Ցանկ") 
print("   POST /orders   - Նոր պատվեր")
print("   GET  /orders   - Բոլոր պատվերները")
print("   GET  /stats    - Վիճակագրություն")
````

# 🆚 FastAPI vs Flask - Համեմատություն

## Ի՞նչ է Flask-ը

**Flask** - Python-ի հանրաճանաչ web framework է, որը գոյություն ունի 2010թ.-ից: Այն "micro-framework" է՝ նվազագույն բաղադրիչներով:

## 🥙 Նույն «Երևանյան Շաուրմա» API-ն Flask-ով

### Flask օրինակ

```python
from flask import Flask, request, jsonify

app = Flask(__name__)

# Տվյալների պահում
menu = {
    "հավով": {"price": 1500, "prep_time": 3},
    "տավարով": {"price": 1800, "prep_time": 4}
}
orders = {}
next_id = 1

@app.route('/menu', methods=['GET'])
def get_menu():
    return jsonify({"menu": menu})

@app.route('/orders', methods=['POST'])
def create_order():
    global next_id
    data = request.get_json()
    
    # Ձեռքով validation
    if not data or 'customer_name' not in data:
        return jsonify({"error": "customer_name պարտադիր է"}), 400
    
    order = {
        "id": next_id,
        "customer_name": data['customer_name'],
        "items": data.get('items', [])
    }
    orders[next_id] = order
    next_id += 1
    
    return jsonify(order), 201

if __name__ == '__main__':
    app.run(debug=True)
```

### FastAPI օրինակ (նույն ֆունկցիոնալություն)

```python
from fastapi import FastAPI
from pydantic import BaseModel

app = FastAPI()

class OrderCreate(BaseModel):  # Ավտոմատ validation
    customer_name: str
    items: list[str] = []

menu = {
    "հավով": {"price": 1500, "prep_time": 3},
    "տավարով": {"price": 1800, "prep_time": 4}
}
orders = {}
next_id = 1

@app.get('/menu')  # Type hints
async def get_menu():  # Async support
    return {"menu": menu}

@app.post('/orders')
async def create_order(order: OrderCreate):  # Ավտոմատ validation
    global next_id
    new_order = {
        "id": next_id,
        "customer_name": order.customer_name,
        "items": order.items
    }
    orders[next_id] = new_order
    next_id += 1
    return new_order
```
