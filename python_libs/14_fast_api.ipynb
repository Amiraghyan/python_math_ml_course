## ü•ô ’ç’´’∂’≠÷Ä’∏’∂ ’¥’∏’ø’•÷Å’∏÷Ç’¥ (Blocking)

```
‘¥’∏÷Ç÷Ñ ‚Üí ’é’°’≥’°’º’∏’≤: "’Ñ’•’Ø ’∑’°’∏÷Ç÷Ä’¥’°’ù ’∞’°’æ’∏’æ, ’≠’∂’±’∏÷Ä’∏’æ, ’°’∂’∏÷Ç’∑’°’∞’∏’ø ’¢’°’∂’ª’°÷Ä’•’≤’•’∂’∏’æ"
’é’°’≥’°’º’∏’≤: "‘º’°’æ, 5 ÷Ä’∏’∫’• ’Ω’∫’°’Ω’•÷Ñ"
[‘¥’∏÷Ç÷Ñ ’Ø’°’∂’£’∂’°’Æ ’Ω’∫’°’Ω’∏÷Ç’¥ ’•÷Ñ 5 ÷Ä’∏’∫’•, ’∏’π’´’∂’π ’π’•÷Ñ ’Ø’°÷Ä’∏’≤ ’°’∂’•’¨]
’é’°’≥’°’º’∏’≤: "‘±’∞’° ’±’•÷Ä ’∑’°’∏÷Ç÷Ä’¥’°’∂"
‘¥’∏÷Ç÷Ñ: "’á’∂’∏÷Ä’∞’°’Ø’°’¨’∏÷Ç’©’µ’∏÷Ç’∂"
[’Ñ’´’°’µ’∂ ’°’µ’™’¥ ’Ø’°÷Ä’∏’≤ ’•÷Ñ ’∑’°÷Ä’∏÷Ç’∂’°’Ø’•’¨ ’∏÷Ç÷Ä’´’∑ ’£’∏÷Ä’Æ’•÷Ä]
```

## üîÑ ‘±’Ω’´’∂’≠÷Ä’∏’∂ ’¥’∏’ø’•÷Å’∏÷Ç’¥ (Non-blocking)

```
‘¥’∏÷Ç÷Ñ ‚Üí ’é’°’≥’°’º’∏’≤: "’Ñ’•’Ø ’∑’°’∏÷Ç÷Ä’¥’°’ù ’∞’°’æ’∏’æ, ’≠’∂’±’∏÷Ä’∏’æ, ’°’∂’∏÷Ç’∑’°’∞’∏’ø ’¢’°’∂’ª’°÷Ä’•’≤’•’∂’∏’æ"
’é’°’≥’°’º’∏’≤: "‘º’°’æ, 5 ÷Ä’∏’∫’• ’Ø’ø÷á’´, ’•’Ω ’±’•’¶ ’Ø’Ø’°’∂’π’•’¥"
[‘¥’∏÷Ç÷Ñ ’∑’°÷Ä’∏÷Ç’∂’°’Ø’∏÷Ç’¥ ’•÷Ñ ’∏÷Ç÷Ä’´’∑ ’£’∏÷Ä’Æ’•÷Ä’ù ’∞’•’º’°’≠’∏’Ω ’≠’∏’Ω’•’¨, ’´’∂’Ω’ø’°’£÷Ä’°’¥ ’∂’°’µ’•’¨]
[4 ÷Ä’∏’∫’• ’°’∂÷Å]
’é’°’≥’°’º’∏’≤: "’Å’•÷Ä ’∑’°’∏÷Ç÷Ä’¥’°’∂ ’∫’°’ø÷Ä’°’Ω’ø ’ß!"
‘¥’∏÷Ç÷Ñ: [’æ’•÷Ä’°’§’°’º’∂’∏÷Ç’¥ ’•÷Ñ] "’á’∂’∏÷Ä’∞’°’Ø’°’¨’∏÷Ç’©’µ’∏÷Ç’∂"
```

**‘±’º’°’æ’•’¨’∏÷Ç’©’µ’∏÷Ç’∂’®** - ‘±’Ω’´’∂’≠÷Ä’∏’∂ ’¥’∏’ø’•÷Å’¥’°’∂ ’§’•’∫÷Ñ’∏÷Ç’¥ ’§’∏÷Ç÷Ñ ’π’•÷Ñ ’™’°’¥’°’∂’°’Ø ’Ø’∏÷Ä÷Å’∂’∏÷Ç’¥ ’Ω’∫’°’Ω’•’¨’´’Ω:

````python
import time
import asyncio
from datetime import datetime

def current_time():
    return datetime.now().strftime("%H:%M:%S")

# ’ç’´’∂’≠÷Ä’∏’∂ (blocking) ’¥’∏’ø’•÷Å’∏÷Ç’¥
def order_shawarma_sync(name, prep_time):
    print(f"{current_time()} - {name}: ’ä’°’ø’æ’´÷Ä’∏÷Ç’¥ ’•’¥ ’∑’°’∏÷Ç÷Ä’¥’°")
    print(f"{current_time()} - ’é’°’≥’°’º’∏’≤: ‘º’°’æ, {prep_time} ’æ’°’µ÷Ä’Ø’µ’°’∂ ’Ø’ø÷á’´")
    
    # ‘≤’¨’∏’Ø’°’æ’∏÷Ä’æ’°’Æ ’Ω’∫’°’Ω’∏÷Ç’¥
    time.sleep(prep_time)
    
    print(f"{current_time()} - ’é’°’≥’°’º’∏’≤: {name}, ’±’•÷Ä ’∑’°’∏÷Ç÷Ä’¥’°’∂ ’∫’°’ø÷Ä’°’Ω’ø ’ß!")
    return f"’á’°’∏÷Ç÷Ä’¥’° {name}-’´ ’∞’°’¥’°÷Ä"

def main_sync():
    print("ü•ô ’ç‘ª’Ü‘Ω’ê’à’Ü ’Ñ’à’è‘µ’ë’à’í’Ñ")
    print("=" * 50)
    
    start_time = time.time()
    
    # ‘µ÷Ä’•÷Ñ ’°’∂’± ’∫’°’ø’æ’´÷Ä’∏÷Ç’¥ ’•’∂
    result1 = order_shawarma_sync("‘±÷Ä’°’¥", 3)
    result2 = order_shawarma_sync("’Ü’°÷Ä’•", 2)
    result3 = order_shawarma_sync("‘º÷á’∏’∂", 4)
    
    end_time = time.time()
    total_time = end_time - start_time
    
    print(f"\n‚è±Ô∏è  ‘∏’∂’§’∞’°’∂’∏÷Ç÷Ä ’™’°’¥’°’∂’°’Ø: {total_time:.1f} ’æ’°’µ÷Ä’Ø’µ’°’∂")
    print("‘≤’∏’¨’∏÷Ä’® ’∫’°’ø’æ’´÷Ä’•’¨ ’•’∂ ’¥’•’Ø ’°’º ’¥’•’Ø, ’Ω’∫’°’Ω’•’¨’∏’æ ’´÷Ä’°÷Ä’°’∂÷Å\n")

# ‘π’•’Ω’ø ’°÷Ä’ø’°’∞’°’µ’ø’•’∂÷Ñ
main_sync()
````

````python
# ‘±’Ω’´’∂’≠÷Ä’∏’∂ (non-blocking) ’¥’∏’ø’•÷Å’∏÷Ç’¥
async def order_shawarma_async(name, prep_time):
    print(f"{current_time()} - {name}: ’ä’°’ø’æ’´÷Ä’∏÷Ç’¥ ’•’¥ ’∑’°’∏÷Ç÷Ä’¥’°")
    print(f"{current_time()} - ’é’°’≥’°’º’∏’≤: ‘º’°’æ {name}, {prep_time} ’æ’°’µ÷Ä’Ø’µ’°’∂ ’Ø’ø÷á’´")
    
    # ‘±’Ω’´’∂’≠÷Ä’∏’∂ ’Ω’∫’°’Ω’∏÷Ç’¥ (’°’µ’¨ ’£’∏÷Ä’Æ’∏’≤’∏÷Ç’©’µ’∏÷Ç’∂’∂’•÷Ä ’Ø’°÷Ä’∏’≤ ’•’∂ ’∑’°÷Ä’∏÷Ç’∂’°’Ø’æ’•’¨)
    await asyncio.sleep(prep_time)
    
    print(f"{current_time()} - ’é’°’≥’°’º’∏’≤: {name}, ’±’•÷Ä ’∑’°’∏÷Ç÷Ä’¥’°’∂ ’∫’°’ø÷Ä’°’Ω’ø ’ß!")
    return f"’á’°’∏÷Ç÷Ä’¥’° {name}-’´ ’∞’°’¥’°÷Ä"

async def main_async():
    print("üöÄ ‘±’ç‘ª’Ü‘Ω’ê’à’Ü ’Ñ’à’è‘µ’ë’à’í’Ñ")
    print("=" * 50)
    
    start_time = time.time()
    
    # ‘≤’∏’¨’∏÷Ä ’∫’°’ø’æ’•÷Ä’∂’•÷Ä’® ’Ω’Ø’Ω’æ’∏÷Ç’¥ ’•’∂ ’∞’°’¥’°’™’°’¥’°’∂’°’Ø’µ’°
    tasks = [
        order_shawarma_async("‘±÷Ä’°’¥", 3),
        order_shawarma_async("’Ü’°÷Ä’•", 2), 
        order_shawarma_async("‘º÷á’∏’∂", 4)
    ]
    
    # ’ç’∫’°’Ω’∏÷Ç’¥ ’•’∂÷Ñ ’¢’∏’¨’∏÷Ä ’∫’°’ø’æ’•÷Ä’∂’•÷Ä’´ ’°’æ’°÷Ä’ø’´’∂
    results = await asyncio.gather(*tasks)
    
    end_time = time.time()
    total_time = end_time - start_time
    
    print(f"\n‚è±Ô∏è  ‘∏’∂’§’∞’°’∂’∏÷Ç÷Ä ’™’°’¥’°’∂’°’Ø: {total_time:.1f} ’æ’°’µ÷Ä’Ø’µ’°’∂")
    print("‘≤’∏’¨’∏÷Ä ’∫’°’ø’æ’•÷Ä’∂’•÷Ä’® ’Ø’°’ø’°÷Ä’æ’•’¨ ’•’∂ ’∞’°’¥’°’™’°’¥’°’∂’°’Ø’µ’°!")
    
    return results

# ‘π’•’Ω’ø ’°÷Ä’ø’°’∞’°’µ’ø’•’∂÷Ñ
await main_async()
````

## üîë ’Ä’´’¥’∂’°’Ø’°’∂ ’¢’°’º’•÷Ä ÷á ’∞’°’Ω’Ø’°÷Å’∏÷Ç’©’µ’∏÷Ç’∂’∂’•÷Ä

- **‘±÷Ä’§’µ’∏÷Ç’∂÷Ñ**: ’ñ’∏÷Ç’∂’Ø÷Å’´’°’∂ ’æ’•÷Ä’°’§’°÷Ä’±’∂’∏÷Ç’¥ ’ß `coroutine` ÷Ö’¢’µ’•’Ø’ø

```python
async def cook_shawarma():
    # ‘±’µ’Ω ÷Ü’∏÷Ç’∂’Ø÷Å’´’°’∂ ’°’Ω’´’∂’≠÷Ä’∏’∂ ’ß
    pass
```

### `await` ’¢’°’º’®

### –ö–ª—é—á–µ–≤—ã–µ –æ—Ç–ª–∏—á–∏—è

| ’ç’´’∂’≠÷Ä’∏’∂ | ‘±’Ω’´’∂’≠÷Ä’∏’∂ |
|---------|-----------|
| `time.sleep()` | `await asyncio.sleep()` |
| ‘≤’¨’∏’Ø’°’æ’∏÷Ä’∏÷Ç’¥ | ’â’´ ’¢’¨’∏’Ø’°’æ’∏÷Ä’∏÷Ç’¥ |
| ’Ä’•÷Ä’©’°’Ø’°’∂ ’Ø’°’ø’°÷Ä’∏÷Ç’¥ | ’Ä’°’¥’°’™’°’¥’°’∂’°’Ø’µ’° ’Ø’°’ø’°÷Ä’∏÷Ç’¥ |
| 9 ’æ’°’µ÷Ä’Ø’µ’°’∂ | 4 ’æ’°’µ÷Ä’Ø’µ’°’∂ |

````python
# üìö ‘≥’∏÷Ä’Æ’∂’°’Ø’°’∂ ÷Ö÷Ä’´’∂’°’Ø’∂’•÷Ä

# 1. ’ä’°÷Ä’¶ ’°’Ω’´’∂’≠÷Ä’∏’∂ ÷Ü’∏÷Ç’∂’Ø÷Å’´’°
async def greet_customer(name, delay):
    print(f"’à’≤’ª’∏÷Ç’µ’∂ {name}! ‘¥’∏÷Ç÷Ä’Ω ’•’Ø’•÷Ñ {delay} ’æ’°’µ÷Ä’Ø’µ’°’∂’´÷Å")
    await asyncio.sleep(delay)
    print(f"{name}, ’±’•÷Ä ’∫’°’ø’æ’•÷Ä’® ’∫’°’ø÷Ä’°’Ω’ø ’ß!")
    return f"‘æ’°’º’°’µ’∏÷Ç’©’µ’∏÷Ç’∂ {name}-’´ ’∞’°’¥’°÷Ä"

# 2. ’Ñ’´ ÷Ñ’°’∂’´ ’°’Ω’´’∂’≠÷Ä’∏’∂ ’£’∏÷Ä’Æ’∏’≤’∏÷Ç’©’µ’∏÷Ç’∂
async def demo_multiple_customers():
    print("üè™ ‘µ÷Ä÷á’°’∂’µ’°’∂ ’á’°’∏÷Ç÷Ä’¥’° - ‘≥’∏÷Ä’Æ’°’Ø’°’¨’∏÷Ç’©’µ’∏÷Ç’∂")
    print("-" * 40)
    
    # ’Ä’°’¥’°’™’°’¥’°’∂’°’Ø’µ’° ’Æ’°’º’°’µ’∏÷Ç’©’µ’∏÷Ç’∂
    customers = [
        greet_customer("‘±’∂’∂’°", 2),
        greet_customer("’é’°’∞’°’£’∂", 1),
        greet_customer("’Ñ’°÷Ä’´’°’¥", 3)
    ]
    
    results = await asyncio.gather(*customers)
    print(f"\n‚úÖ ‘æ’°’º’°’µ’∏÷Ç’©’µ’∏÷Ç’∂ ’°’æ’°÷Ä’ø’æ’°’Æ: {len(results)} ’∞’°’≥’°’≠’∏÷Ä’§")
    return results

# ‘ø’°’ø’°÷Ä’•’¨
await demo_multiple_customers()
````

````python
# üö® ’ç’≠’°’¨’∂’•÷Ä’´ ’¥’∑’°’Ø’∏÷Ç’¥ ’°’Ω’´’∂’≠÷Ä’∏’∂ ÷Ü’∏÷Ç’∂’Ø÷Å’´’°’∂’•÷Ä’∏÷Ç’¥

async def prepare_special_shawarma(name, is_available=True):
    print(f"’ä’°’ø÷Ä’°’Ω’ø’∏÷Ç’¥ ’•’¥ ’∞’°’ø’∏÷Ç’Ø ’∑’°’∏÷Ç÷Ä’¥’° {name}-’´ ’∞’°’¥’°÷Ä...")
    await asyncio.sleep(2)
    
    if not is_available:
        raise Exception(f"‘ø’∂’•÷Ä’•÷Ñ {name}, ’∞’°’ø’∏÷Ç’Ø ’¢’°’≤’°’§÷Ä’´’π’∂’•÷Ä’® ’æ’•÷Ä’ª’°÷Å’•’¨ ’•’∂!")
    
    print(f"‚úÖ ’Ä’°’ø’∏÷Ç’Ø ’∑’°’∏÷Ç÷Ä’¥’° ’∫’°’ø÷Ä’°’Ω’ø ’ß {name}-’´ ’∞’°’¥’°÷Ä")
    return f"’Ä’°’ø’∏÷Ç’Ø ’∑’°’∏÷Ç÷Ä’¥’° - {name}"

async def handle_orders_with_errors():
    print("üçΩÔ∏è  ’Ä’°’ø’∏÷Ç’Ø ’∫’°’ø’æ’•÷Ä’∂’•÷Ä’´ ’¥’∑’°’Ø’∏÷Ç’¥")
    print("-" * 30)
    
    orders = [
        prepare_special_shawarma("‘≥’°’£’´’Ø", True),   # ’Ä’°’ª’∏’≤’°’Ø’°’∂
        prepare_special_shawarma("’Ü’æ’°÷Ä’§", False),  # ’ç’≠’°’¨
        prepare_special_shawarma("‘±÷Ä’¥’•’∂", True),   # ’Ä’°’ª’∏’≤’°’Ø’°’∂
    ]
    
    # ’Ñ’∑’°’Ø’∏÷Ç’¥’ù ’Ω’≠’°’¨’∂’•÷Ä’´ ’∞’•’ø
    results = []
    for order in orders:
        try:
            result = await order
            results.append(result)
        except Exception as e:
            print(f"‚ùå ’ç’≠’°’¨: {e}")
            results.append(None)
    
    successful_orders = [r for r in results if r is not None]
    print(f"\nüìä ‘±’¥÷É’∏÷É’∏÷Ç’¥: {len(successful_orders)}/{len(results)} ’∫’°’ø’æ’•÷Ä ’∞’°’ª’∏’≤’æ’•÷Å")
    
    return results

# ‘π’•’Ω’ø
await handle_orders_with_errors()
````

````python
# üîÑ HTTP Methods - ‘µ÷Ä÷á’°’∂’µ’°’∂ ’á’°’∏÷Ç÷Ä’¥’° ÷Ö÷Ä’´’∂’°’Ø’∂’•÷Ä’∏’æ

# HTTP Methods-’´ ÷Å’∏÷Ç÷Å’°’§÷Ä’∏÷Ç’©’µ’∏÷Ç’∂
def demonstrate_http_methods():
    """HTTP ’¥’•’©’∏’§’∂’•÷Ä’´ ’¢’°÷Å’°’ø÷Ä’∏÷Ç’©’µ’∏÷Ç’∂ ’º’•’Ω’ø’∏÷Ä’°’∂’´ ÷Ö÷Ä’´’∂’°’Ø’∏’æ"""
    
    print("üåê HTTP METHODS - ‘µ’ê‘µ’é‘±’Ü’Ö‘±’Ü ’á‘±’à’í’ê’Ñ‘±")
    print("=" * 50)
    
    methods = {
        "GET": {
            "description": "’è’•’≤’•’Ø’∏÷Ç’©’µ’∏÷Ç’∂ ’Ω’ø’°’∂’°’¨ (’π÷É’∏’≠’•’¨ ’∏’π’´’∂’π)",
            "example": "’ë’°’∂’Ø’´ ’§’´’ø’∏÷Ç’¥, ’∫’°’ø’æ’•÷Ä’´ ’Ø’°÷Ä’£’°’æ’´’≥’°’Ø’´ ’Ω’ø’∏÷Ç’£’∏÷Ç’¥",
            "shawarma_case": "¬´‘ª’û’∂’π ’∑’°’∏÷Ç÷Ä’¥’° ’∏÷Ç’∂’•÷Ñ¬ª, ¬´‘ª’û’∂’π ’£’´’∂ ’ß¬ª",
            "safe": True,
            "idempotent": True
        },
        "POST": {
            "description": "’Ü’∏÷Ä ’¢’°’∂ ’Ω’ø’•’≤’Æ’•’¨",
            "example": "’Ü’∏÷Ä ’∫’°’ø’æ’•÷Ä ’ø’°’¨",
            "shawarma_case": "¬´’à÷Ç’¶’∏÷Ç’¥ ’•’¥ ’¥’•’Ø ’∞’°’æ’∏’æ ’∑’°’∏÷Ç÷Ä’¥’° ’∫’°’ø’æ’´÷Ä’•’¨¬ª",
            "safe": False,
            "idempotent": False
        },
        "PUT": {
            "description": "‘±’¥’¢’∏’≤’ª’∏÷Ç’©’µ’°’¥’¢ ÷É’∏’≠’°÷Ä’´’∂’•’¨/’©’°÷Ä’¥’°÷Å’∂’•’¨",
            "example": "’ä’°’ø’æ’•÷Ä’® ’°’¥’¢’∏’≤’ª’∏÷Ç’©’µ’°’¥’¢ ÷É’∏’≠’•’¨",
            "shawarma_case": "¬´’â’•’≤’°÷Ä’Ø’•÷Ñ ’∞’°’æ’®, ’∏÷Ç’¶’∏÷Ç’¥ ’•’¥ ’ø’°’æ’°÷Ä’∏’æ¬ª",
            "safe": False,
            "idempotent": True
        },
        "PATCH": {
            "description": "’Ñ’°’Ω’∂’°’Ø’´’∏÷Ä’•’∂ ’©’°÷Ä’¥’°÷Å’∂’•’¨",
            "example": "’Ñ’´’°’µ’∂ ’¥’•’Ø ’¢’°’≤’°’§÷Ä’´’π ÷É’∏’≠’•’¨",
            "shawarma_case": "¬´‘Ω’∂’§÷Ä’∏÷Ç’¥ ’•’¥ ’°’º’°’∂÷Å ’Ω’∏’≠’´¬ª",
            "safe": False,
            "idempotent": False
        },
        "DELETE": {
            "description": "’ã’∂’ª’•’¨",
            "example": "’ä’°’ø’æ’•÷Ä’® ’π’•’≤’°÷Ä’Ø’•’¨",
            "shawarma_case": "¬´’â’•’≤’°÷Ä’Ø’∏÷Ç’¥ ’•’¥ ’´’¥ ’∫’°’ø’æ’•÷Ä’®¬ª",
            "safe": False,
            "idempotent": True
        }
    }
    
    for method, info in methods.items():
        print(f"\nüî∏ {method}")
        print(f"   üìù ’Ü’Ø’°÷Ä’°’£÷Ä’∏÷Ç’©’µ’∏÷Ç’∂: {info['description']}")
        print(f"   ü•ô ’á’°’∏÷Ç÷Ä’¥’° ÷Ö÷Ä’´’∂’°’Ø: {info['shawarma_case']}")
        print(f"   ‚ö° ‘±’∂’æ’ø’°’∂’£: {'‘±’µ’∏' if info['safe'] else '’à’π'}")
        print(f"   üîÑ Idempotent: {'‘±’µ’∏' if info['idempotent'] else '’à’π'}")
        
    print(f"\nüí° Idempotent = ’Ü’∏÷Ç’µ’∂ ’£’∏÷Ä’Æ’∏’≤’∏÷Ç’©’µ’∏÷Ç’∂’® ’Ø÷Ä’Ø’∂’•’¨’´’Ω ’∂’∏÷Ç’µ’∂ ’°÷Ä’§’µ’∏÷Ç’∂÷Ñ’®")

# ’ë’∏÷Ç÷Å’°’§÷Ä’∏÷Ç’©’µ’∏÷Ç’∂
demonstrate_http_methods()
````

````python
# üõ°Ô∏è Error Handling - ‘≥’∏÷Ä’Æ’∂’°’Ø’°’∂ ÷Ö÷Ä’´’∂’°’Ø’∂’•÷Ä

from fastapi import HTTPException
import random

# ’ä’°÷Ä’¶ ’Ω’≠’°’¨’∂’•÷Ä’´ ÷Å’∏÷Ç÷Å’°’§÷Ä’∏÷Ç’©’µ’∏÷Ç’∂
def demonstrate_status_codes():
    """HTTP status codes-’´ ÷Å’∏÷Ç÷Å’°’§÷Ä’∏÷Ç’©’µ’∏÷Ç’∂"""
    
    print("üö¶ HTTP STATUS CODES - ‘µ’ê‘µ’é‘±’Ü’Ö‘±’Ü ’á‘±’à’í’ê’Ñ‘±")
    print("=" * 55)
    
    scenarios = [
        {
            "situation": "’Ä’°’≥’°’≠’∏÷Ä’§’® ’∞’°÷Ä÷Å’∂’∏÷Ç’¥ ’ß ÷Å’°’∂’Ø’®",
            "code": 200,
            "message": "OK - ’ë’°’∂’Ø’® ’¢’°÷Ä’•’∞’°’ª’∏’≤ ’∏÷Ç’≤’°÷Ä’Ø’æ’•÷Å"
        },
        {
            "situation": "’Ü’∏÷Ä ’∫’°’ø’æ’•÷Ä ’ß ’Ω’ø’•’≤’Æ’æ’•’¨",
            "code": 201,
            "message": "Created - ’ä’°’ø’æ’•÷Ä’® ’∞’°’ª’∏’≤’∏÷Ç’©’µ’°’¥’¢ ’Ω’ø’•’≤’Æ’æ’•÷Å"
        },
        {
            "situation": "’Ä’°÷Ä÷Å’∂’∏÷Ç’¥ ’•’∂ ’π’£’∏’µ’∏÷Ç’©’µ’∏÷Ç’∂ ’∏÷Ç’∂’•÷Å’∏’≤ ’∏÷Ç’ø’•’¨’´÷Ñ",
            "code": 404,
            "message": "Not Found - ‘±’µ’§’∫’´’Ω’´ ’∏÷Ç’ø’•’¨’´÷Ñ ’π’∏÷Ç’∂’•’∂÷Ñ"
        },
        {
            "situation": "’ä’°’ø’æ’´÷Ä’∏÷Ç’¥ ’•’∂ ÷É’°’Ø ’™’°’¥’•÷Ä’´’∂",
            "code": 403,
            "message": "Forbidden - ’å’•’Ω’ø’∏÷Ä’°’∂’® ÷É’°’Ø ’ß"
        },
        {
            "situation": "’ç’≠’°’¨ ’ø’æ’µ’°’¨’∂’•÷Ä ’•’∂ ’∏÷Ç’≤’°÷Ä’Ø’•’¨",
            "code": 400,
            "message": "Bad Request - ’â’∞’°’Ω’Ø’°÷Å’° ’±’•÷Ä ’∫’°’ø’æ’•÷Ä’®"
        },
        {
            "situation": "‘Ω’∏’∞’°’∂’∏÷Å’∏÷Ç’¥ ’∞÷Ä’§’•’∞ ’ß",
            "code": 500,
            "message": "Internal Server Error - ’Ü’•÷Ä÷Ñ’´’∂ ’≠’∂’§’´÷Ä"
        }
    ]
    
    for i, scenario in enumerate(scenarios, 1):
        status_emoji = "‚úÖ" if 200 <= scenario["code"] < 300 else "‚ùå" if 400 <= scenario["code"] < 600 else "üîÑ"
        print(f"\n{i}. {scenario['situation']}")
        print(f"   {status_emoji} {scenario['code']} - {scenario['message']}")

# ’ç’≠’°’¨’∂’•÷Ä’´ simulation
class ShawarmaError(Exception):
    """’á’°’∏÷Ç÷Ä’¥’°-’´ ’Ω’≠’°’¨’∂’•÷Ä’´ class"""
    pass

def simulate_order_errors(order_item: str):
    """’ä’°’ø’æ’•÷Ä’´ ’Ω’≠’°’¨’∂’•÷Ä’´ simulation"""
    
    # ’è’°÷Ä’¢’•÷Ä ’Ω’≠’°’¨’∂’•÷Ä’´ ÷Å’∏÷Ç÷Å’°’§÷Ä’∏÷Ç’©’µ’∏÷Ç’∂
    error_scenarios = {
        "’ø’°’æ’°÷Ä’∏’æ": {"available": True, "prep_time": 5},
        "’∞’°’æ’∏’æ": {"available": True, "prep_time": 3},
        "’¢’°’∂’ª’°÷Ä’•’≤’•’∂’∏’æ": {"available": False, "error": "‘≤’°’∂’ª’°÷Ä’•’≤’•’∂’® ’æ’•÷Ä’ª’°÷Å’•’¨ ’ß"},
        "’±’Ø’∂’∏’æ": {"available": False, "error": "’Å’∏÷Ç’Ø ’π’∏÷Ç’∂’•’∂÷Ñ"},
        "’∞’°’ø’∏÷Ç’Ø": {"available": True, "prep_time": 8, "special_error": "’Ä’°’ø’∏÷Ç’Ø ’¢’°’≤’°’§÷Ä’´’π’® ’π’Ø’°"}
    }
    
    if order_item not in error_scenarios:
        raise HTTPException(status_code=404, detail=f"'{order_item}' ’∑’°’∏÷Ç÷Ä’¥’° ’π’∏÷Ç’∂’•’∂÷Ñ")
    
    scenario = error_scenarios[order_item]
    
    if not scenario["available"]:
        raise HTTPException(status_code=422, detail=scenario["error"])
    
    if "special_error" in scenario and random.choice([True, False]):
        raise HTTPException(status_code=500, detail=scenario["special_error"])
    
    return {"message": f"{order_item} ’∑’°’∏÷Ç÷Ä’¥’° ’∫’°’ø÷Ä’°’Ω’ø ’Ø’¨’´’∂’´ {scenario['prep_time']} ÷Ä’∏’∫’•’∏÷Ç’¥", 
            "prep_time": scenario["prep_time"]}

# ‘π’•’Ω’ø
print("üìã ’ç‘Ω‘±‘º’Ü‘µ’ê‘ª ’ë’à’í’ë‘±‘¥’ê’à’í‘π’Ö’à’í’Ü:")
demonstrate_status_codes()

print("\n" + "="*60)
print("üß™ ’ä‘±’è’é‘µ’ê’Ü‘µ’ê‘ª ‘π‘µ’ç’è:")

test_orders = ["’∞’°’æ’∏’æ", "’¢’°’∂’ª’°÷Ä’•’≤’•’∂’∏’æ", "’±’Ø’∂’∏’æ", "’Ø’°’ø’æ’∏’æ"]

for order in test_orders:
    try:
        result = simulate_order_errors(order)
        print(f"‚úÖ {order}: {result['message']}")
    except HTTPException as e:
        print(f"‚ùå {order}: {e.status_code} - {e.detail}")
    except Exception as e:
        print(f"üí• {order}: ‘±’∂’Ω’∫’°’Ω’•’¨’´ ’Ω’≠’°’¨ - {e}")
````

````python
# üöÄ ‘±’¥’¢’∏’≤’ª’°’Ø’°’∂ FastAPI ÷Ö÷Ä’´’∂’°’Ø - ‘µ÷Ä÷á’°’∂’µ’°’∂ ’á’°’∏÷Ç÷Ä’¥’°

# ’ë’°’∂’Ø’°’µ’´’∂ ’°’µ’Ω ÷Ö÷Ä’´’∂’°’Ø’® ’°’∑’≠’°’ø’°÷Å’∂’•’¨’∏÷Ç ’∞’°’¥’°÷Ä ’∫’•’ø÷Ñ ’ß ’ø’•’≤’°’§÷Ä’•’¨ FastAPI:
# pip install fastapi uvicorn

from pydantic import BaseModel, validator
from typing import Optional, List
from datetime import datetime
import asyncio

# --- ’è’é’Ö‘±‘º’Ü‘µ’ê‘ª ’Ñ’à‘¥‘µ‘º’Ü‘µ’ê ---

class ShawarmaItem(BaseModel):
    name: str
    price: int
    available: bool = True
    prep_time: int  # ÷Ä’∏’∫’•’∂’•÷Ä

class OrderCreate(BaseModel):
    customer_name: str
    items: List[str]
    special_requests: Optional[str] = ""
    
    @validator('customer_name')
    def validate_name(cls, v):
        if len(v.strip()) < 2:
            raise ValueError('‘±’∂’∏÷Ç’∂’® ’∫’•’ø÷Ñ ’ß ’¨’´’∂’´ ’∂’æ’°’¶’°’£’∏÷Ç’µ’∂’® 2 ’ø’°’º')
        return v.strip()

class Order(BaseModel):
    id: int
    customer_name: str
    items: List[str]
    total_price: int
    status: str
    created_at: datetime
    estimated_time: int

# --- "’è’é’Ö‘±‘º’Ü‘µ’ê‘ª ‘≤‘±‘∂‘±" ---
menu_items = {
    "’∞’°’æ’∏’æ": ShawarmaItem(name="’∞’°’æ’∏’æ", price=1500, prep_time=3),
    "’ø’°’æ’°÷Ä’∏’æ": ShawarmaItem(name="’ø’°’æ’°÷Ä’∏’æ", price=1800, prep_time=4),
    "’¢’°’∂’ª’°÷Ä’•’≤’•’∂’∏’æ": ShawarmaItem(name="’¢’°’∂’ª’°÷Ä’•’≤’•’∂’∏’æ", price=1200, prep_time=2),
    "’∞’°’ø’∏÷Ç’Ø": ShawarmaItem(name="’∞’°’ø’∏÷Ç’Ø", price=2200, prep_time=6)
}

orders_storage = {}
next_order_id = 1

# --- ’ï‘∫‘±’Ü‘¥‘±‘ø ’ñ’à’í’Ü‘ø’ë‘ª‘±’Ü‘µ’ê ---

async def validate_menu_items(items: List[str]) -> None:
    """’ç’ø’∏÷Ç’£’•’¨’ù ’°÷Ä’§’µ’∏÷Ñ ’∫’°’ø’æ’´÷Ä’æ’°’Æ ’°’∫÷Ä’°’∂÷Ñ’∂’•÷Ä’® ’Ø’°’∂ ÷Å’°’∂’Ø’∏÷Ç’¥"""
    for item in items:
        if item not in menu_items:
            raise HTTPException(
                status_code=404, 
                detail=f"'{item}' ’∑’°’∏÷Ç÷Ä’¥’° ’¥’•’∂÷Ñ ’π’∏÷Ç’∂’•’∂÷Ñ: ‘±’º’Ø’° ’ø’°÷Ä’¢’•÷Ä’°’Ø’∂’•÷Ä’ù {list(menu_items.keys())}"
            )
        if not menu_items[item].available:
            raise HTTPException(
                status_code=422,
                detail=f"'{item}' ’∑’°’∏÷Ç÷Ä’¥’°’∂ ’™’°’¥’°’∂’°’Ø’°’æ’∏÷Ä’°’∫’•’Ω ’¥’°’ø’π’•’¨’´ ’π’ß"
            )

def calculate_order_total(items: List[str]) -> tuple[int, int]:
    """’Ä’°’∑’æ’•’¨ ’∫’°’ø’æ’•÷Ä’´ ’£’∏÷Ç’¥’°÷Ä’® ÷á ’∫’°’ø÷Ä’°’Ω’ø’¥’°’∂ ’™’°’¥’°’∂’°’Ø’®"""
    total_price = sum(menu_items[item].price for item in items)
    total_time = max(menu_items[item].prep_time for item in items)  # ‘±’¥’•’∂’°’•÷Ä’Ø’°÷Ä ’™’°’¥’°’∂’°’Ø
    return total_price, total_time

# --- FASTAPI ENDPOINTS SIMULATION ---

async def simulate_get_menu():
    """GET /menu - ’ë’°’∂’Ø’´ ’Ω’ø’°÷Å’∏÷Ç’¥"""
    return {
        "status": "success",
        "menu": {name: {"price": item.price, "available": item.available} 
                for name, item in menu_items.items()}
    }

async def simulate_create_order(order_data: OrderCreate):
    """POST /orders - ’Ü’∏÷Ä ’∫’°’ø’æ’•÷Ä ’Ω’ø’•’≤’Æ’∏÷Ç’¥"""
    global next_order_id
    
    # Validation
    await validate_menu_items(order_data.items)
    
    # Calculation
    total_price, prep_time = calculate_order_total(order_data.items)
    
    # Create order
    new_order = Order(
        id=next_order_id,
        customer_name=order_data.customer_name,
        items=order_data.items,
        total_price=total_price,
        status="’£’∏÷Ä’Æ’®’∂’©’°÷Å",
        created_at=datetime.now(),
        estimated_time=prep_time
    )
    
    orders_storage[next_order_id] = new_order
    order_id = next_order_id
    next_order_id += 1
    
    # Simulate cooking
    print(f"ü•ô ’ä’°’ø÷Ä’°’Ω’ø’∏÷Ç’¥ ’•’¥ ’∫’°’ø’æ’•÷Ä #{order_id} ({order_data.customer_name})")
    await asyncio.sleep(1)  # Simulation
    
    return {
        "status": "created",
        "order": new_order,
        "message": f"’ä’°’ø’æ’•÷Ä #{order_id} ’Ω’ø’•’≤’Æ’æ’•÷Å: ’ä’°’ø÷Ä’°’Ω’ø ’Ø’¨’´’∂’´ {prep_time} ÷Ä’∏’∫’•’∏÷Ç’¥"
    }

async def simulate_get_order(order_id: int):
    """GET /orders/{order_id} - ’Ñ’•’Ø ’∫’°’ø’æ’•÷Ä’´ ’ø’æ’µ’°’¨’∂’•÷Ä"""
    if order_id not in orders_storage:
        raise HTTPException(status_code=404, detail=f"’ä’°’ø’æ’•÷Ä #{order_id} ’π’´ ’£’ø’∂’æ’•’¨")
    
    return {"status": "success", "order": orders_storage[order_id]}

async def simulate_update_order(order_id: int, new_items: List[str]):
    """PUT /orders/{order_id} - ’ä’°’ø’æ’•÷Ä’® ÷É’∏’≠’•’¨"""
    if order_id not in orders_storage:
        raise HTTPException(status_code=404, detail=f"’ä’°’ø’æ’•÷Ä #{order_id} ’π’´ ’£’ø’∂’æ’•’¨")
    
    order = orders_storage[order_id]
    if order.status != "’£’∏÷Ä’Æ’®’∂’©’°÷Å":
        raise HTTPException(status_code=422, detail="’ä’°’ø÷Ä’°’Ω’ø ’∫’°’ø’æ’•÷Ä’® ’π’´ ’Ø’°÷Ä’•’¨’´ ÷É’∏’≠’•’¨")
    
    await validate_menu_items(new_items)
    total_price, prep_time = calculate_order_total(new_items)
    
    order.items = new_items
    order.total_price = total_price
    order.estimated_time = prep_time
    
    return {"status": "updated", "order": order}

async def simulate_cancel_order(order_id: int):
    """DELETE /orders/{order_id} - ’ä’°’ø’æ’•÷Ä’® ’π’•’≤’°÷Ä’Ø’•’¨"""
    if order_id not in orders_storage:
        raise HTTPException(status_code=404, detail=f"’ä’°’ø’æ’•÷Ä #{order_id} ’π’´ ’£’ø’∂’æ’•’¨")
    
    order = orders_storage[order_id]
    if order.status == "’∫’°’ø÷Ä’°’Ω’ø":
        raise HTTPException(status_code=422, detail="’ä’°’ø÷Ä’°’Ω’ø ’∫’°’ø’æ’•÷Ä’® ’π’´ ’Ø’°÷Ä’•’¨’´ ’π’•’≤’°÷Ä’Ø’•’¨")
    
    del orders_storage[order_id]
    return {"status": "cancelled", "message": f"’ä’°’ø’æ’•÷Ä #{order_id} ’π’•’≤’°÷Ä’Ø’æ’•÷Å"}

# --- ’è‘µ’ç’è‘±’é’à’ê’à’í’Ñ ---

async def test_api_endpoints():
    """API endpoints-’´ ’©’•’Ω’ø’°’æ’∏÷Ä’∏÷Ç’¥"""
    print("üß™ FASTAPI ENDPOINTS ‘π‘µ’ç’è")
    print("=" * 50)
    
    try:
        # 1. GET Menu
        print("\n1Ô∏è‚É£ GET /menu")
        menu_result = await simulate_get_menu()
        print(f"‚úÖ ’ë’°’∂’Ø’® ’Ω’ø’°÷Å’æ’•÷Å: {len(menu_result['menu'])} ’ø’°÷Ä’¢’•÷Ä’°’Ø")
        
        # 2. POST Order
        print("\n2Ô∏è‚É£ POST /orders")
        order_data = OrderCreate(
            customer_name="‘±÷Ä’°’¥ ’ä’•’ø÷Ä’∏’Ω’µ’°’∂", 
            items=["’∞’°’æ’∏’æ", "’ø’°’æ’°÷Ä’∏’æ"],
            special_requests="‘±’º’°’∂÷Å ’Ω’∏’≠’´"
        )
        order_result = await simulate_create_order(order_data)
        order_id = order_result["order"].id
        print(f"‚úÖ {order_result['message']}")
        
        # 3. GET Specific Order
        print("\n3Ô∏è‚É£ GET /orders/{order_id}")
        get_result = await simulate_get_order(order_id)
        print(f"‚úÖ ’ä’°’ø’æ’•÷Ä #{order_id} ’£’ø’∂’æ’•÷Å")
        
        # 4. PUT Update Order
        print("\n4Ô∏è‚É£ PUT /orders/{order_id}")
        update_result = await simulate_update_order(order_id, ["’∞’°’ø’∏÷Ç’Ø"])
        print(f"‚úÖ ’ä’°’ø’æ’•÷Ä’® ’©’°÷Ä’¥’°÷Å’æ’•÷Å")
        
        # 5. ’ì’∏÷Ä’± - ’Ω’≠’°’¨ ’∫’°’ø’æ’•÷Ä
        print("\n5Ô∏è‚É£ POST /orders (’Ω’≠’°’¨)")
        try:
            bad_order = OrderCreate(customer_name="‘≥’°’£’´’Ø", items=["’Ø’°’ø’æ’∏’æ"])
            await simulate_create_order(bad_order)
        except HTTPException as e:
            print(f"‚ùå ‘±’Ø’∂’Ø’°’¨’æ’∏’≤ ’Ω’≠’°’¨: {e.status_code} - {e.detail}")
        
        # 6. DELETE Cancel Order
        print("\n6Ô∏è‚É£ DELETE /orders/{order_id}")
        cancel_result = await simulate_cancel_order(order_id)
        print(f"‚úÖ {cancel_result['message']}")
        
    except Exception as e:
        print(f"üí• ‘±’∂’Ω’∫’°’Ω’•’¨’´ ’Ω’≠’°’¨: {e}")

# ‘π’•’Ω’ø
await test_api_endpoints()
````

### üéØ ‘µ’û÷Ä’¢ ÷Ö’£’ø’°’£’∏÷Ä’Æ’•’¨

#### Flask-’® ’®’∂’ø÷Ä’•÷Ñ’ù ’•’©’•
- ‚úÖ ’Ä’°÷Ä’¥’°÷Ä ’•÷Ñ minimal approach-’´ ’∞’•’ø
- ‚úÖ ‘æ’°’∂’∏’© ’•÷Ñ Flask ecosystem-’´’∂
- ‚úÖ ’à÷Ç’∂’•÷Ñ ’£’∏’µ’∏÷Ç’©’µ’∏÷Ç’∂ ’∏÷Ç’∂’•÷Å’∏’≤ Flask app
- ‚úÖ ’à÷Ç’¶’∏÷Ç’¥ ’•÷Ñ ’°’¥’•’∂ ’´’∂’π ’æ’•÷Ä’°’∞’Ω’Ø’•’¨

````python
# üß™ API-’∂’•÷Ä’´ ’©’•’Ω’ø’°’æ’∏÷Ä’∏÷Ç’¥

import requests
import json
import sys
import os

def test_api_endpoint(base_url, endpoint, method="GET", data=None):
    """API endpoint-’´ ’©’•’Ω’ø’°’æ’∏÷Ä’¥’°’∂ ÷Ü’∏÷Ç’∂’Ø÷Å’´’°"""
    url = f"{base_url}{endpoint}"
    
    try:
        if method == "GET":
            response = requests.get(url, timeout=5)
        elif method == "POST":
            headers = {"Content-Type": "application/json"}
            response = requests.post(url, json=data, headers=headers, timeout=5)
        else:
            return f"‚ùå ‘±’∂’°’Ø’∂’Ø’°’¨ method: {method}"
        
        if response.status_code == 200:
            return f"‚úÖ {method} {endpoint}: {response.status_code}"
        elif response.status_code == 201:
            return f"‚úÖ {method} {endpoint}: {response.status_code} - Created"
        else:
            return f"‚ö†Ô∏è {method} {endpoint}: {response.status_code} - {response.text[:100]}"
    
    except requests.exceptions.ConnectionError:
        return f"‚ùå {method} {endpoint}: ’ç’•÷Ä’æ’•÷Ä’® ’π’´ ’°’∑’≠’°’ø’∏÷Ç’¥"
    except requests.exceptions.Timeout:
        return f"‚è∞ {method} {endpoint}: Timeout"
    except Exception as e:
        return f"üí• {method} {endpoint}: {str(e)}"

def test_apis():
    """FastAPI ÷á Flask API-’∂’•÷Ä’´ ’©’•’Ω’ø’°’æ’∏÷Ä’∏÷Ç’¥"""
    
    print("üß™ API ENDPOINTS ‘π‘µ’ç’è")
    print("=" * 50)
    
    # API bases
    fastapi_base = "http://localhost:8000"
    flask_base = "http://localhost:5000"
    
    # Test data
    test_order = {
        "customer_name": "‘±÷Ä’°’¥ ‘π’•’Ω’ø’°’Ø’°’∂",
        "items": ["’∞’°’æ’∏’æ", "’¢’°’∂’ª’°÷Ä’•’≤’•’∂’∏’æ"],
        "special_requests": "‘±’º’°’∂÷Å ’Ω’∏’≠’´"
    }
    
    print("\nüöÄ FastAPI ’©’•’Ω’ø (Port 8000)")
    print("-" * 30)
    
    fastapi_tests = [
        (fastapi_base, "/", "GET", None),
        (fastapi_base, "/menu", "GET", None),
        (fastapi_base, "/orders", "POST", test_order),
        (fastapi_base, "/orders/1", "GET", None),
        (fastapi_base, "/orders", "GET", None),
    ]
    
    for base, endpoint, method, data in fastapi_tests:
        result = test_api_endpoint(base, endpoint, method, data)
        print(f"  {result}")
    
    print("\nüå∂Ô∏è Flask ’©’•’Ω’ø (Port 5000)")
    print("-" * 30)
    
    flask_tests = [
        (flask_base, "/", "GET", None),
        (flask_base, "/menu", "GET", None),
        (flask_base, "/orders", "POST", test_order),
        (flask_base, "/orders/1", "GET", None),
        (flask_base, "/orders", "GET", None),
    ]
    
    for base, endpoint, method, data in flask_tests:
        result = test_api_endpoint(base, endpoint, method, data)
        print(f"  {result}")
    
    print(f"\nüí° ’Ü’Ø’°’ø’°’º’∏÷Ç’¥:")
    print(f"   - ‘≥’∏÷Ä’Æ’°÷Ä’Ø’•÷Ñ API-’∂’•÷Ä’® ’°’º’°’∂’±’´’∂ terminal-’∂’•÷Ä’∏÷Ç’¥")
    print(f"   - FastAPI: cd apis && python fastapi_shawarma.py")
    print(f"   - Flask:   cd apis && python flask_shawarma.py")

# ‘π’•’Ω’ø
test_apis()
````

````python
# üîß FastAPI ’ø’•’≤’°’§÷Ä’∏÷Ç’¥ ÷á ’Ø’°’º’∏÷Ç÷Å’æ’°’Æ÷Ñ

# 1. ’è’•’≤’°’§÷Ä’∏÷Ç’¥ - Terminal-’∏÷Ç’¥ ’£’∏÷Ä’Æ’°÷Ä’Ø’•’¨:
# pip install fastapi uvicorn python-multipart

# 2. ’ä’°÷Ä’ø’°’§’´÷Ä imports
try:
    from fastapi import FastAPI, HTTPException, Depends, Query, Path
    from fastapi.responses import JSONResponse
    from pydantic import BaseModel, Field
    from typing import Optional, List, Dict, Any
    from datetime import datetime, timedelta
    from enum import Enum
    import asyncio
    import json
    print("‚úÖ FastAPI imports ’¢’°÷Ä’•’∞’°’ª’∏’≤ ’¢’•’º’∂’°’æ’∏÷Ä’æ’•÷Å’´’∂")
except ImportError as e:
    print(f"‚ùå Import Error: {e}")
    print("üí° ‘Ω’∏÷Ä’∞’∏÷Ç÷Ä’§: pip install fastapi uvicorn")

# 3. FastAPI application ’Ω’ø’•’≤’Æ’∏÷Ç’¥
app = FastAPI(
    title="ü•ô ‘µ÷Ä÷á’°’∂’µ’°’∂ ’á’°’∏÷Ç÷Ä’¥’° API",
    description="’å’•’Ω’ø’∏÷Ä’°’∂’´ ’∫’°’ø’æ’•÷Ä’°’Ø’¥’°’∂ ’∞’°’¥’°’Ø’°÷Ä’£",
    version="1.0.0",
    docs_url="/docs",  # Swagger UI
    redoc_url="/redoc"  # ReDoc alternative docs
)

print("üöÄ FastAPI application ’Ω’ø’•’≤’Æ’æ’•÷Å")

# 4. Enum classes
class ShawarmaType(str, Enum):
    CHICKEN = "’∞’°’æ’∏’æ"
    BEEF = "’ø’°’æ’°÷Ä’∏’æ"
    VEGGIE = "’¢’°’∂’ª’°÷Ä’•’≤’•’∂’∏’æ"
    SPECIAL = "’∞’°’ø’∏÷Ç’Ø"
    FISH = "’±’Ø’∂’∏’æ"

class OrderStatus(str, Enum):
    PENDING = "’Ω’∫’°’Ω’æ’∏’≤"
    CONFIRMED = "’∞’°’Ω’ø’°’ø’æ’°’Æ"
    PREPARING = "’∫’°’ø÷Ä’°’Ω’ø’æ’∏÷Ç’¥"
    READY = "’∫’°’ø÷Ä’°’Ω’ø"
    DELIVERED = "’°’º’°÷Ñ’æ’°’Æ"
    CANCELLED = "’π’•’≤’°÷Ä’Ø’æ’°’Æ"

# 5. Pydantic Models (v2 compatible)
class MenuItem(BaseModel):
    name: str = Field(..., description="’à÷Ç’ø’•’¨’´÷Ñ’´ ’°’∂’æ’°’∂’∏÷Ç’©’µ’∏÷Ç’∂’®")
    price: int = Field(..., gt=0, description="‘≥’´’∂’® ’§÷Ä’°’¥’∏’æ")
    description: str = Field("", description="’Ü’Ø’°÷Ä’°’£÷Ä’∏÷Ç’©’µ’∏÷Ç’∂’®")
    available: bool = Field(True, description="’Ä’°’Ω’°’∂’•’¨’´’∏÷Ç’©’µ’∏÷Ç’∂’®")
    prep_time: int = Field(..., gt=0, le=30, description="’ä’°’ø÷Ä’°’Ω’ø’¥’°’∂ ’™’°’¥’°’∂’°’Ø’® ÷Ä’∏’∫’•’∂’•÷Ä’∏’æ")
    category: str = Field("shawarma", description="‘ø’°’ø’•’£’∏÷Ä’´’°’∂")
    
    model_config = {
        "json_schema_extra": {
            "example": {
                "name": "’∞’°’æ’∏’æ ’∑’°’∏÷Ç÷Ä’¥’°",
                "price": 1500,
                "description": "’Ä’°’æ’´ ’¥’´’Ω, ’¢’°’∂’ª’°÷Ä’•’≤’•’∂, ’Ω’∏’∏÷Ç’Ω",
                "available": True,
                "prep_time": 5,
                "category": "shawarma"
            }
        }
    }

class OrderItem(BaseModel):
    item_name: str
    quantity: int = Field(..., gt=0, le=10)
    special_requests: Optional[str] = ""

class OrderCreate(BaseModel):
    customer_name: str = Field(..., min_length=2, max_length=50)
    phone: str = Field(..., pattern=r"^\+374\d{8}$")
    items: List[OrderItem] = Field(..., min_length=1)
    delivery_address: Optional[str] = None
    payment_method: str = Field("cash", pattern="^(cash|card|online)$")

class OrderResponse(BaseModel):
    id: int
    customer_name: str
    phone: str
    items: List[OrderItem]
    total_price: int
    status: OrderStatus
    created_at: datetime
    estimated_delivery: Optional[datetime] = None

print("üìù ’è’æ’µ’°’¨’∂’•÷Ä’´ ’¥’∏’§’•’¨’∂’•÷Ä’® ’∫’°’ø÷Ä’°’Ω’ø ’•’∂")
````

````python
# üóÑÔ∏è "’è’æ’µ’°’¨’∂’•÷Ä’´ ’¢’°’¶’°" ÷á ’∞’´’¥’∂’°’Ø’°’∂ ’£’∏÷Ä’Æ’°’º’∏÷Ç’µ’©’∂’•÷Ä

# ’ä’°÷Ä’¶ in-memory "database"
class ShawarmaDatabase:
    def __init__(self):
        self.menu: Dict[str, MenuItem] = {}
        self.orders: Dict[int, OrderResponse] = {}
        self.next_order_id = 1
        self._init_menu()
    
    def _init_menu(self):
        """’ë’°’∂’Ø’´ ’ø’æ’µ’°’¨’∂’•÷Ä’´ ’´’∂’´÷Å’´’°’¨’´’¶’°÷Å’´’°"""
        menu_items = [
            MenuItem(
                name="’∞’°’æ’∏’æ ’∑’°’∏÷Ç÷Ä’¥’°",
                price=1500,
                description="’Ä÷Ä’°’∫’°÷Ä’°’Ø’æ’°’Æ ’∞’°’æ’´ ’¥’´’Ω, ’¥’°’∂÷Ä ’Ø’ø÷Ä’°’Æ ’¢’°’∂’ª’°÷Ä’•’≤’•’∂, ’Ω’∏’∏÷Ç’Ω",
                prep_time=5,
                category="’∑’°’∏÷Ç÷Ä’¥’°"
            ),
            MenuItem(
                name="’ø’°’æ’°÷Ä’∏’æ ’∑’°’∏÷Ç÷Ä’¥’°", 
                price=1800,
                description="’î’´’π ’∫’°’ø÷Ä’°’Ω’ø’æ’°’Æ ’ø’°’æ’°÷Ä’´ ’¥’´’Ω, ’¢’°’∂’ª’°÷Ä’•’≤’•’∂, ’∞’°’ø’∏÷Ç’Ø ’Ω’∏’∏÷Ç’Ω",
                prep_time=6,
                category="’∑’°’∏÷Ç÷Ä’¥’°"
            ),
            MenuItem(
                name="’¢’°’∂’ª’°÷Ä’•’≤’•’∂’∏’æ ’∑’°’∏÷Ç÷Ä’¥’°",
                price=1200,
                description="’è’°÷Ä’¢’•÷Ä ’¢’°’∂’ª’°÷Ä’•’≤’•’∂, ’∞’∏÷Ç’¥’∏÷Ç’Ω, ’Ω’∏’∏÷Ç’Ω",
                prep_time=4,
                category="’∑’°’∏÷Ç÷Ä’¥’°"
            ),
            MenuItem(
                name="’∞’°’ø’∏÷Ç’Ø ’∑’°’∏÷Ç÷Ä’¥’°",
                price=2200,
                description="’Ä’°’æ+’ø’°’æ’°÷Ä, ’∞’°’ø’∏÷Ç’Ø ’Ω’∏’∏÷Ç’Ω, ’Ø’°’Ø’°’π",
                prep_time=8,
                category="’∑’°’∏÷Ç÷Ä’¥’°"
            ),
            MenuItem(
                name="’Ø’∏’¨’°",
                price=400,
                description="Coca-Cola 0.5’¨",
                prep_time=1,
                category="’®’¥’∫’•’¨’´÷Ñ"
            ),
            MenuItem(
                name="’°’µ÷Ä’°’∂",
                price=300,
                description="’è’°’∂’°’Ø’°’∂ ’°’≠’Ø’°’Æ’´÷Ä ’º’´’°’µ’´’∂",
                prep_time=1,
                category="’®’¥’∫’•’¨’´÷Ñ"
            )
        ]
        
        for item in menu_items:
            self.menu[item.name] = item
    
    def get_menu(self) -> Dict[str, MenuItem]:
        """’ë’°’∂’Ø’® ’Ω’ø’°’∂’°’¨"""
        return self.menu
    
    def get_available_menu(self) -> Dict[str, MenuItem]:
        """’Ñ’´’°’µ’∂ ’∞’°’Ω’°’∂’•’¨’´ ’°’∫÷Ä’°’∂÷Ñ’∂’•÷Ä’®"""
        return {name: item for name, item in self.menu.items() if item.available}
    
    def create_order(self, order_data: OrderCreate) -> OrderResponse:
        """’Ü’∏÷Ä ’∫’°’ø’æ’•÷Ä ’Ω’ø’•’≤’Æ’•’¨"""
        # ’ç’ø’∏÷Ç’£’•’¨ ’°’∫÷Ä’°’∂÷Ñ’∂’•÷Ä’´ ’°’º’Ø’°’µ’∏÷Ç’©’µ’∏÷Ç’∂’®
        for item in order_data.items:
            if item.item_name not in self.menu:
                raise HTTPException(
                    status_code=404,
                    detail=f"'{item.item_name}' ÷Å’°’∂’Ø’∏÷Ç’¥ ’π’Ø’°"
                )
            if not self.menu[item.item_name].available:
                raise HTTPException(
                    status_code=422,
                    detail=f"'{item.item_name}' ’™’°’¥’°’∂’°’Ø’°’æ’∏÷Ä’°’∫’•’Ω ’¥’°’ø’π’•’¨’´ ’π’ß"
                )
        
        # ‘≥’∏÷Ç’¥’°÷Ä’´ ’∞’°’∑’æ’°÷Ä’Ø
        total_price = sum(
            self.menu[item.item_name].price * item.quantity 
            for item in order_data.items
        )
        
        # ‘±’æ’°÷Ä’ø’¥’°’∂ ’™’°’¥’°’∂’°’Ø’´ ’£’∂’°’∞’°’ø’∏÷Ç’¥
        max_prep_time = max(
            self.menu[item.item_name].prep_time 
            for item in order_data.items
        )
        estimated_delivery = datetime.now() + timedelta(minutes=max_prep_time + 10)
        
        # ’ä’°’ø’æ’•÷Ä ’Ω’ø’•’≤’Æ’•’¨
        order = OrderResponse(
            id=self.next_order_id,
            customer_name=order_data.customer_name,
            phone=order_data.phone,
            items=order_data.items,
            total_price=total_price,
            status=OrderStatus.PENDING,
            created_at=datetime.now(),
            estimated_delivery=estimated_delivery
        )
        
        self.orders[self.next_order_id] = order
        self.next_order_id += 1
        
        return order
    
    def get_order(self, order_id: int) -> OrderResponse:
        """’ä’°’ø’æ’•÷Ä ’£’ø’∂’•’¨"""
        if order_id not in self.orders:
            raise HTTPException(
                status_code=404,
                detail=f"’ä’°’ø’æ’•÷Ä #{order_id} ’π’´ ’£’ø’∂’æ’•’¨"
            )
        return self.orders[order_id]
    
    def update_order_status(self, order_id: int, new_status: OrderStatus) -> OrderResponse:
        """’ä’°’ø’æ’•÷Ä’´ ’Ø’°÷Ä’£’°’æ’´’≥’°’Ø ÷É’∏’≠’•’¨"""
        order = self.get_order(order_id)
        order.status = new_status
        return order
    
    def cancel_order(self, order_id: int) -> bool:
        """’ä’°’ø’æ’•÷Ä ’π’•’≤’°÷Ä’Ø’•’¨"""
        order = self.get_order(order_id)
        if order.status in [OrderStatus.DELIVERED, OrderStatus.CANCELLED]:
            raise HTTPException(
                status_code=422,
                detail=f"’ä’°’ø’æ’•÷Ä #{order_id} ’∞’∂’°÷Ä’°’æ’∏÷Ä ’π’ß ’π’•’≤’°÷Ä’Ø’•’¨: ‘ø’°÷Ä’£’°’æ’´’≥’°’Ø’ù {order.status.value}"
            )
        
        order.status = OrderStatus.CANCELLED
        return True

# ‘≥’¨’∏’¢’°’¨ database instance
db = ShawarmaDatabase()

print("üóÑÔ∏è ’è’æ’µ’°’¨’∂’•÷Ä’´ ’¢’°’¶’°’∂ ’∫’°’ø÷Ä’°’Ω’ø ’ß")
print(f"üìã ’ë’°’∂’Ø’∏÷Ç’¥ {len(db.menu)} ’°’∫÷Ä’°’∂÷Ñ ’Ø’°")
print(f"üõí ’ä’°’ø’æ’•÷Ä’∂’•÷Ä: {len(db.orders)}")
````

````python
# üõ£Ô∏è FastAPI Routes (Endpoints)

# ROOT endpoint
@app.get("/", tags=["‘≥’¨’≠’°’æ’∏÷Ä"])
async def root():
    """API-’´ ’£’¨’≠’°’æ’∏÷Ä ’ß’ª"""
    return {
        "message": "ü•ô ‘≤’°÷Ä’´ ’£’°’¨’∏÷Ç’Ω’ø ‘µ÷Ä÷á’°’∂’µ’°’∂ ’á’°’∏÷Ç÷Ä’¥’° API",
        "version": "1.0.0",
        "docs": "/docs",
        "status": "’Ø’•’∂’§’°’∂’´"
    }

# MENU endpoints
@app.get("/menu", response_model=Dict[str, MenuItem], tags=["’ë’°’∂’Ø"])
async def get_menu(
    available_only: bool = Query(False, description="’Ñ’´’°’µ’∂ ’∞’°’Ω’°’∂’•’¨’´ ’°’∫÷Ä’°’∂÷Ñ’∂’•÷Ä’®")
):
    """’ë’°’∂’Ø’´ ’Ω’ø’°÷Å’∏÷Ç’¥"""
    if available_only:
        return db.get_available_menu()
    return db.get_menu()

@app.get("/menu/{item_name}", response_model=MenuItem, tags=["’ë’°’∂’Ø"])
async def get_menu_item(item_name: str = Path(..., description="‘±’∫÷Ä’°’∂÷Ñ’´ ’°’∂’æ’°’∂’∏÷Ç’©’µ’∏÷Ç’∂’®")):
    """’Ñ’•’Ø ’°’∫÷Ä’°’∂÷Ñ’´ ’¥’°’∂÷Ä’°’¥’°’Ω’∂’•÷Ä"""
    if item_name not in db.menu:
        raise HTTPException(
            status_code=404,
            detail=f"'{item_name}' ÷Å’°’∂’Ø’∏÷Ç’¥ ’π’´ ’£’ø’∂’æ’•’¨"
        )
    return db.menu[item_name]

# ORDER endpoints
@app.post("/orders", response_model=OrderResponse, status_code=201, tags=["’ä’°’ø’æ’•÷Ä’∂’•÷Ä"])
async def create_order(order: OrderCreate):
    """’Ü’∏÷Ä ’∫’°’ø’æ’•÷Ä ’Ω’ø’•’≤’Æ’•’¨"""
    try:
        new_order = db.create_order(order)
        
        # ‘±’Ω’´’∂’≠÷Ä’∏’∂ ’∫’°’ø÷Ä’°’Ω’ø’¥’°’∂ ’Ω’´’¥’∏÷Ç’¨’µ’°÷Å’´’°
        asyncio.create_task(simulate_order_preparation(new_order.id))
        
        return new_order
    except Exception as e:
        if isinstance(e, HTTPException):
            raise e
        raise HTTPException(status_code=500, detail=f"’Ü’•÷Ä÷Ñ’´’∂ ’Ω’≠’°’¨: {str(e)}")

@app.get("/orders/{order_id}", response_model=OrderResponse, tags=["’ä’°’ø’æ’•÷Ä’∂’•÷Ä"])
async def get_order(order_id: int = Path(..., description="’ä’°’ø’æ’•÷Ä’´ ’∞’°’¥’°÷Ä’®")):
    """’ä’°’ø’æ’•÷Ä’´ ’Ø’°÷Ä’£’°’æ’´’≥’°’Ø ’Ω’ø’∏÷Ç’£’•’¨"""
    return db.get_order(order_id)

@app.get("/orders", response_model=List[OrderResponse], tags=["’ä’°’ø’æ’•÷Ä’∂’•÷Ä"])
async def get_all_orders(
    status: Optional[OrderStatus] = Query(None, description="‘ø’°÷Ä’£’°’æ’´’≥’°’Ø’∏’æ ÷Ü’´’¨’ø÷Ä"),
    limit: int = Query(10, ge=1, le=100, description="‘±’º’°’æ’•’¨’°’£’∏÷Ç’µ’∂ ÷Ñ’°’∂’°’Ø")
):
    """‘≤’∏’¨’∏÷Ä ’∫’°’ø’æ’•÷Ä’∂’•÷Ä’®"""
    orders = list(db.orders.values())
    
    if status:
        orders = [order for order in orders if order.status == status]
    
    # ’é’•÷Ä’ª’´’∂ ’∫’°’ø’æ’•÷Ä’∂’•÷Ä’® ’æ’•÷Ä÷á’∏÷Ç’¥
    orders.sort(key=lambda x: x.created_at, reverse=True)
    
    return orders[:limit]

@app.put("/orders/{order_id}/status", response_model=OrderResponse, tags=["’ä’°’ø’æ’•÷Ä’∂’•÷Ä"])
async def update_order_status(
    order_id: int = Path(..., description="’ä’°’ø’æ’•÷Ä’´ ’∞’°’¥’°÷Ä’®"),
    new_status: OrderStatus = Query(..., description="’Ü’∏÷Ä ’Ø’°÷Ä’£’°’æ’´’≥’°’Ø")
):
    """’ä’°’ø’æ’•÷Ä’´ ’Ø’°÷Ä’£’°’æ’´’≥’°’Ø ÷É’∏’≠’•’¨"""
    return db.update_order_status(order_id, new_status)

@app.delete("/orders/{order_id}", tags=["’ä’°’ø’æ’•÷Ä’∂’•÷Ä"])
async def cancel_order(order_id: int = Path(..., description="’ä’°’ø’æ’•÷Ä’´ ’∞’°’¥’°÷Ä’®")):
    """’ä’°’ø’æ’•÷Ä ’π’•’≤’°÷Ä’Ø’•’¨"""
    success = db.cancel_order(order_id)
    if success:
        return {"message": f"’ä’°’ø’æ’•÷Ä #{order_id} ’π’•’≤’°÷Ä’Ø’æ’•÷Å", "status": "success"}
    
# STATISTICS endpoint
@app.get("/stats", tags=["’é’´’≥’°’Ø’°’£÷Ä’∏÷Ç’©’µ’∏÷Ç’∂"])
async def get_statistics():
    """’Ä’´’¥’∂’°’Ø’°’∂ ’æ’´’≥’°’Ø’°’£÷Ä’∏÷Ç’©’µ’∏÷Ç’∂"""
    orders = list(db.orders.values())
    
    total_orders = len(orders)
    total_revenue = sum(order.total_price for order in orders if order.status != OrderStatus.CANCELLED)
    
    status_counts = {}
    for status in OrderStatus:
        status_counts[status.value] = len([o for o in orders if o.status == status])
    
    popular_items = {}
    for order in orders:
        for item in order.items:
            if item.item_name not in popular_items:
                popular_items[item.item_name] = 0
            popular_items[item.item_name] += item.quantity
    
    return {
        "total_orders": total_orders,
        "total_revenue": total_revenue,
        "status_breakdown": status_counts,
        "popular_items": dict(sorted(popular_items.items(), key=lambda x: x[1], reverse=True)[:5])
    }

# ’ï’™’°’∂’§’°’Ø ÷Ü’∏÷Ç’∂’Ø÷Å’´’° - ’∫’°’ø’æ’•÷Ä’´ ’∫’°’ø÷Ä’°’Ω’ø’¥’°’∂ ’Ω’´’¥’∏÷Ç’¨’µ’°÷Å’´’°
async def simulate_order_preparation(order_id: int):
    """’ä’°’ø’æ’•÷Ä’´ ’∫’°’ø÷Ä’°’Ω’ø’¥’°’∂ ’°’Ω’´’∂’≠÷Ä’∏’∂ ’Ω’´’¥’∏÷Ç’¨’µ’°÷Å’´’°"""
    try:
        # ’Ä’°’Ω’ø’°’ø’•’¨
        await asyncio.sleep(2)
        db.update_order_status(order_id, OrderStatus.CONFIRMED)
        
        # ’ä’°’ø÷Ä’°’Ω’ø’•’¨
        await asyncio.sleep(3)
        db.update_order_status(order_id, OrderStatus.PREPARING)
        
        # ’ä’°’ø÷Ä’°’Ω’ø
        order = db.get_order(order_id)
        prep_times = [db.menu[item.item_name].prep_time for item in order.items]
        total_prep_time = max(prep_times) if prep_times else 5
        
        await asyncio.sleep(total_prep_time)
        db.update_order_status(order_id, OrderStatus.READY)
        
        # ‘±’º’°÷Ñ’•’¨ (÷Ö÷Ä’´’∂’°’Ø)
        await asyncio.sleep(10)
        db.update_order_status(order_id, OrderStatus.DELIVERED)
        
    except Exception as e:
        print(f"’ç’≠’°’¨ ’∫’°’ø’æ’•÷Ä #{order_id} ’∫’°’ø÷Ä’°’Ω’ø’•’¨’´’Ω: {e}")

print("üõ£Ô∏è FastAPI routes ’∫’°’ø÷Ä’°’Ω’ø ’•’∂")
print("üìä Endpoints:")
print("   GET  /         - ‘≥’¨’≠’°’æ’∏÷Ä ’ß’ª")
print("   GET  /menu     - ’ë’°’∂’Ø") 
print("   POST /orders   - ’Ü’∏÷Ä ’∫’°’ø’æ’•÷Ä")
print("   GET  /orders   - ‘≤’∏’¨’∏÷Ä ’∫’°’ø’æ’•÷Ä’∂’•÷Ä’®")
print("   GET  /stats    - ’é’´’≥’°’Ø’°’£÷Ä’∏÷Ç’©’µ’∏÷Ç’∂")
````

# üÜö FastAPI vs Flask - ’Ä’°’¥’•’¥’°’ø’∏÷Ç’©’µ’∏÷Ç’∂

### Flask ÷Ö÷Ä’´’∂’°’Ø

```python

app = Flask(__name__)

# ’è’æ’µ’°’¨’∂’•÷Ä’´ ’∫’°’∞’∏÷Ç’¥
menu = {
    "’∞’°’æ’∏’æ": {"price": 1500, "prep_time": 3},
    "’ø’°’æ’°÷Ä’∏’æ": {"price": 1800, "prep_time": 4}
}
```
