```python
fig, axs = plt.subplots(2, 2)  # Creates a grid of 2 rows and 2 columns
axs[0, 0].plot([1, 2, 3, 4], [1, 4, 9, 16])  # Plots on the subplot in the first row and first column
axs[0, 1].bar([1, 2, 3, 4], [1, 4, 9, 16])  # Plots a bar chart on the subplot in the first row and second column
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, sharex=True, sharey=True)  # The subplots share both x and y axes
axs[0, 0].plot([1, 2, 3, 4], [1, 4, 9, 16])
axs[0, 1].bar([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
```

## մինի գործնական (դասին չարեցինք մենք, բայց դե կոդը թողեմ)

```python
import streamlit as st

st.title('Երբ կդառնանք միլիոնատեր')


def calculate_returns(starting_amount, years_to_invest, additional_contributions, annual_rate_of_return, compounding_frequency):
    total_amount = starting_amount
    compound_rate = 1 + annual_rate_of_return / 100
    compounding_periods = {
        "Annually": 1,
        "Half-yearly": 2,
        "Quarterly": 4,
        "Monthly": 12
    }
    periods = int(years_to_invest) * compounding_periods[compounding_frequency]
    returns = []

    for year in range(1, int(years_to_invest) + 1):
        total_amount += additional_contributions
        total_amount *= compound_rate
        returns.append(total_amount)

    return returns

starting_amount = st.number_input('Սկզբնական գումար', value=0)
monthly_contribution = st.number_input('Ամսական գումար', value=0, step=1000)
interest_rate = st.number_input('Տոկոսադրույք', value=0.0, step=0.1)
years = st.number_input('Տարիներ', value=0, step=1)
frequency = st.selectbox('Հաճախականություն',  ['Monthly', 'Quarterly', 'Half-yearly', 'Annually'])

if st.button("հաշվել"):
    incomes = calculate_returns(starting_amount, years, monthly_contribution, interest_rate, frequency)
    st.subheader("Եկամուտներ")
    st.write(incomes)

    st.subheader("Գրաֆիկ")
    st.line_chart(incomes)

    st.dataframe(incomes)

    st.download_button(file_name='incomes.txt', label='Ներբեռնել', data=str(incomes))
```
