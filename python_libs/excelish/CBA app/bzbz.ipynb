```python
import os
import pandas as pd
import plotly.express as px 
```

```python
URL = "https://www.cba.am/stat/stat_data_arm/Loans_by_regions_arm.xlsx"
```

```python
df_himnakan = pd.read_excel(URL)
df_dynamic = pd.read_excel(URL, sheet_name="Դինամիկ շարք")
```

```python
df_himnakan
```

```python
df_himnakan.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
himnakan_name = df_himnakan.iloc[0, 0]
himnakan_date = df_himnakan.iloc[3, 0]
print(himnakan_name, himnakan_date)
```

```python
df_himnakan = df_himnakan.iloc[4:, :] 
df_himnakan 
```

```python
cols = df_himnakan.iloc[0]
cols.fillna("Anun", inplace=True)
cols 
```

```python
df_himnakan.columns = cols
df_himnakan
```

```python
# drop first row
df_himnakan.drop(4, inplace=True)
```

```python
df_himnakan
```

```python
df_himnakan.reset_index(drop=True, inplace=True)
```

```python
# remove index column
df_himnakan.columns.name = None

df_himnakan
```

```python
nshum = df_himnakan.iloc[-1:,0 ]
```

```python
# drop last 2 rows
df_himnakan.drop(df_himnakan.tail(2).index, inplace=True)
```

```python
df_himnakan = df_himnakan.set_index("Anun").T
```

```python
df_himnakan["Ընդհանուր"] = df_himnakan.sum(axis=1)
df_himnakan
```

```python
df_himnakan.rename(columns={"index": "marz"}, inplace=True)
```

```python
df_himnakan.reset_index(inplace=True)
```

```python
df_himnakan
```

```python
df_long = pd.melt(df_himnakan, id_vars=["index"], var_name="loan_type", value_name="amount")
```

```python
df_long.rename(columns={"index": "region"}, inplace=True)
```

```python
# select only first 5 loan types
df_long = df_long[df_long["loan_type"].isin(df_long["loan_type"].unique()[:5])]
df_long
```

```python
df_long.loan_type = df_long.loan_type.apply(lambda x: x.split(" ")[0])
```

```python
# Stacked area plot to see the contribution of each region to loan amounts
fig = px.area(df_long, x='loan_type', y='amount', color='region',
              title='Stacked Loan Amounts by Region')
fig.show()
```

```python
# Pie chart to see the percentage of total loans per region
df_total_by_region = df_long.groupby('region')['amount'].sum().reset_index()


# donut chart
fig = px.pie(df_total_by_region, values='amount', names='region', title='Total Loans by Region')
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

```python
# Heatmap to visualize loan amounts for each loan type and region
fig = px.density_heatmap(df_long, x='loan_type', y='region', z='amount',
                         title='Loan Amounts Heatmap by Type and Region')
fig.show()
```
