```python
df_long = pd.melt(df_h, id_vars="Վարկի տեսակ", var_name="region", value_name="amount")
```

```python
df_long.rename(columns={"Վարկի տեսակ": "loan_type"}, inplace=True)
```

```python
df_long.loan_type.iloc[-1]
```

```python
df_long.loan_type = df_long.loan_type.apply(lambda x: x.strip())
df_long.region = df_long.region.apply(lambda x: x.strip())


df_long.loan_type = df_long.loan_type.apply(lambda x: x[0])
```

## Df հիմնական plots

### Distribution for each region for fixed loan type

```python
loan = "Սպառողական վարկեր"

df_f = df_long[df_long["loan_type"] == loan]
df_f 
```

```python
px.bar(df_long, x="region", y="amount", facet_row="loan_type",title=f"{loan} by region")
```

### Distribution for each loan type for fixed region

```python
region = "01․Երևան"

df_f = df_long[df_long["region"] == region]
df_f 
```

```python
px.line(df_f, x="loan_type", y="amount", title=f"{region} loans by type")
```

```python
mean_amount_by_region = df_long.groupby("loan_type").amount.mean().reset_index()
mean_amount_by_region
```

```python
#Todo 2 plotly in one figure
fig1 = px.area(df_long, x="loan_type", y="amount")

fig2 = px.scatter(mean_amount_by_region, x="loan_type", y="amount")

fig1.add_trace(fig2.data[0])
```

```python
import plotly.express as px
import plotly.graph_objects as go

# Sample data
x = ['Category A', 'Category B', 'Category C']
y_bar = [10, 15, 13]
y_scatter = [11, 9, 16]

# Create a bar plot with Plotly Express
fig = px.bar(x=x, y=y_bar, labels={'x':'Category', 'y':'Value'}, title="Bar Plot and Scatter Plot Overlay")

# Add a scatter plot trace to the same figure
fig.add_trace(px.scatter(x=x, y=y_scatter))

# Show the figure
fig.show()
```
